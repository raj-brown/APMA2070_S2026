# Implementation of NAdam


```python
# Source: https://github.com/rwightman/pytorch-image-models/blob/master/timm/optim/nadam.py
import math

import torch
from torch.optim.optimizer import Optimizer

class Nadam(Optimizer):
    """Implements Nadam algorithm (a variant of Adam based on Nesterov momentum).
    It has been proposed in `Incorporating Nesterov Momentum into Adam`__.
    Arguments:
        params (iterable): iterable of parameters to optimize or dicts defining
            parameter groups
        lr (float, optional): learning rate (default: 2e-3)
        betas (Tuple[float, float], optional): coefficients used for computing
            running averages of gradient and its square
        eps (float, optional): term added to the denominator to improve
            numerical stability (default: 1e-8)
        weight_decay (float, optional): weight decay (L2 penalty) (default: 0)
        schedule_decay (float, optional): momentum schedule decay (default: 4e-3)
    __ http://cs229.stanford.edu/proj2015/054_report.pdf
    __ http://www.cs.toronto.edu/~fritz/absps/momentum.pdf
        Originally taken from: https://github.com/pytorch/pytorch/pull/1408
        NOTE: Has potential issues but does work well on some problems.
    """

    def __init__(self, params, lr=2e-3, betas=(0.9, 0.999), eps=1e-8,
                 weight_decay=0, schedule_decay=4e-3):
        if not 0.0 <= lr:
            raise ValueError("Invalid learning rate: {}".format(lr))
        defaults = dict(
            lr=lr, betas=betas, eps=eps, weight_decay=weight_decay, schedule_decay=schedule_decay)
        super(Nadam, self).__init__(params, defaults)

    @torch.no_grad()
    def step(self, closure=None):
        """Performs a single optimization step.
        Arguments:
            closure (callable, optional): A closure that reevaluates the model
                and returns the loss.
        """
        loss = None
        if closure is not None:
            with torch.enable_grad():
                loss = closure()

        for group in self.param_groups:
            for p in group['params']:
                if p.grad is None:
                    continue
                grad = p.grad
                state = self.state[p]

                # State initialization
                if len(state) == 0:
                    state['step'] = 0
                    state['m_schedule'] = 1.
                    state['exp_avg'] = torch.zeros_like(p)
                    state['exp_avg_sq'] = torch.zeros_like(p)

                # Warming momentum schedule
                m_schedule = state['m_schedule']
                schedule_decay = group['schedule_decay']
                exp_avg, exp_avg_sq = state['exp_avg'], state['exp_avg_sq']
                beta1, beta2 = group['betas']
                eps = group['eps']
                state['step'] += 1
                t = state['step']
                bias_correction2 = 1 - beta2 ** t

                if group['weight_decay'] != 0:
                    grad = grad.add(p, alpha=group['weight_decay'])

                momentum_cache_t = beta1 * (1. - 0.5 * (0.96 ** (t * schedule_decay)))
                momentum_cache_t_1 = beta1 * (1. - 0.5 * (0.96 ** ((t + 1) * schedule_decay)))
                m_schedule_new = m_schedule * momentum_cache_t
                m_schedule_next = m_schedule * momentum_cache_t * momentum_cache_t_1
                state['m_schedule'] = m_schedule_new

                # Decay the first and second moment running average coefficient
                exp_avg.mul_(beta1).add_(grad, alpha=1. - beta1)
                exp_avg_sq.mul_(beta2).addcmul_(grad, grad, value=1. - beta2)

                denom = (exp_avg_sq.sqrt() / math.sqrt(bias_correction2)).add_(eps)
                p.addcdiv_(grad, denom, value=-group['lr'] * (1. - momentum_cache_t) / (1. - m_schedule_new))
                p.addcdiv_(exp_avg, denom, value=-group['lr'] * momentum_cache_t_1 / (1. - m_schedule_next))

        return loss
```

# Function Approximation Using NADAM

```python
# ============================================
# NAdam Optimization (Neural regression) + GIF + Loss plot + Final prediction plot
# ============================================

import numpy as np
import imageio
import torch
import matplotlib.pyplot as plt

%matplotlib inline
torch.manual_seed(1234)

print("NAdam Optimization")

# --------------------------------------------
# Target function (piecewise)
#   x < 0  : 5 + sum of sines
#   x >= 0 : cos(10x)
# --------------------------------------------
def fun_x(x):
    if x < 0.0:
        return 5.0 + np.sin(x) + np.sin(2.0 * x) + np.sin(3.0 * x) + np.sin(4.0 * x)
    else:
        return np.cos(10.0 * x)

# --------------------------------------------
# Build training data
# --------------------------------------------
x_in_l = np.linspace(-np.pi, -1.0e-3, 201)
x_in_r = np.linspace(0.0, np.pi, 501)
x_in = np.concatenate((x_in_l, x_in_r), axis=0)
y_in = np.array([fun_x(i) for i in x_in])

# Plot training function + points
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x_in, y_in, 'k-', label="True Function")
ax.plot(x_in, y_in, '.b', label="Training Points")
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y', fontsize=20)
ax.legend()
plt.show()

# Convert to torch tensors (shape: [N, 1])
x = torch.from_numpy(x_in.reshape(-1, 1)).float()
y = torch.from_numpy(y_in.reshape(-1, 1)).float()

# --------------------------------------------
# Model, loss, optimizer
# --------------------------------------------
loss_fn = torch.nn.MSELoss()
Niter = 500

Net = torch.nn.Sequential(
    torch.nn.Linear(1, 100),
    torch.nn.ReLU(),
    torch.nn.Linear(100, 1)
)

# NAdam optimizer (PyTorch)
optimizer = torch.optim.NAdam(Net.parameters(), lr=1e-3)

# --------------------------------------------
# GIF setup: we render frames every `frame_every` iterations
# --------------------------------------------
fig_anim, ax_anim = plt.subplots(figsize=(15, 10))
image_list = []

# Loss history (store floats)
nadam_loss = []

frame_every = 20     # capture a GIF frame every 20 iterations
print_every = 50     # print progress every 50 iterations

# --------------------------------------------
# Training loop
# --------------------------------------------
for it in range(Niter):
    # Forward pass
    y_pred = Net(x)

    # Compute MSE loss
    loss = loss_fn(y_pred, y)

    # Backprop + update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Log loss as float
    nadam_loss.append(loss.item())

    # Print occasionally (printing every iteration is slow)
    if (it == 0) or ((it + 1) % print_every == 0):
        print(f"Iter: {it + 1:03}/{Niter}  Loss: {loss.item():.5e}")

    # ----------------------------------------
    # Capture frames for GIF occasionally
    # ----------------------------------------
    if it % frame_every == 0:
        with torch.no_grad():
            y_pred_np = Net(x).detach().cpu().numpy()

        ax_anim.cla()
        ax_anim.set_xlabel('x', fontsize=20)
        ax_anim.set_ylabel('y', fontsize=20)
        ax_anim.set_xlim(-4, 4)
        ax_anim.set_ylim(-1.5, 7.0)

        ax_anim.scatter(x_in, y_in, label="Training Data")
        ax_anim.plot(x_in, y_pred_np, 'k--', lw=2, label="NN Prediction")

        ax_anim.text(-3.5, 6.2, f"NAdam Optimizer (iter={it})", fontsize=16)
        ax_anim.text(-3.5, 5.7, f"Loss = {loss.item():.6f}", fontsize=16)
        ax_anim.legend()

        fig_anim.canvas.draw()
        frame = np.frombuffer(fig_anim.canvas.tostring_rgb(), dtype=np.uint8)
        frame = frame.reshape(fig_anim.canvas.get_width_height()[::-1] + (3,))
        image_list.append(frame)

# --------------------------------------------
# Save GIF
# --------------------------------------------
gif_path = "./NAdam_func.gif"
imageio.mimsave(gif_path, image_list, fps=5)
print(f"Saved GIF to: {gif_path}")

# ============================================
# Plot 1: Loss curve
# ============================================
fig_loss, ax_loss = plt.subplots(figsize=(12, 7))
ep = np.arange(1, Niter + 1)

ax_loss.semilogy(ep, nadam_loss, 'r-', lw=2, label="NAdam Train Loss")
ax_loss.set_xlabel("Iterations")
ax_loss.set_ylabel("MSE Loss (log scale)")
ax_loss.set_title("Training Loss with NAdam")
ax_loss.legend()
plt.show()

# ============================================
# Plot 2: Final prediction vs true function
# ============================================
with torch.no_grad():
    y_final = Net(x).detach().cpu().numpy()

fig_pred, ax_pred = plt.subplots(figsize=(12, 7))
ax_pred.plot(x_in, y_in, "k-", label="True Function")
ax_pred.scatter(x_in, y_in, label="Training Points")
ax_pred.plot(x_in, y_final, "b-", lw=2, label="Final NN Prediction (NAdam)")

ax_pred.set_xlabel("x")
ax_pred.set_ylabel("y")
ax_pred.set_title("Final Fit after NAdam Optimization")
ax_pred.legend()
plt.show()
```

```python
# ============================================
# RMSprop Optimization (Neural regression) + GIF + Loss plot + Final prediction plot
# ============================================

import numpy as np
import imageio
import torch
import matplotlib.pyplot as plt

%matplotlib inline
torch.manual_seed(1234)

print("RMSprop Optimization")

# --------------------------------------------
# Target function (piecewise)
#   x < 0  : 5 + sum of sines
#   x >= 0 : cos(10x)
# --------------------------------------------
def fun_x(x):
    if x < 0.0:
        return 5.0 + np.sin(x) + np.sin(2.0 * x) + np.sin(3.0 * x) + np.sin(4.0 * x)
    else:
        return np.cos(10.0 * x)

# --------------------------------------------
# Build training data
# --------------------------------------------
x_in_l = np.linspace(-np.pi, -1.0e-3, 201)
x_in_r = np.linspace(0.0, np.pi, 501)
x_in = np.concatenate((x_in_l, x_in_r), axis=0)
y_in = np.array([fun_x(i) for i in x_in])

# Plot training data
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x_in, y_in, 'k-', label="True Function")
ax.plot(x_in, y_in, '.b', label="Training Points")
ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)
ax.legend()
plt.show()

# Convert to tensors (shape [N,1])
x = torch.from_numpy(x_in.reshape(-1, 1)).float()
y = torch.from_numpy(y_in.reshape(-1, 1)).float()

# --------------------------------------------
# Model + loss + optimizer
# --------------------------------------------
Net = torch.nn.Sequential(
    torch.nn.Linear(1, 100),
    torch.nn.ReLU(),
    torch.nn.Linear(100, 1)
)

loss_fn = torch.nn.MSELoss()
Niter = 500

# RMSprop optimizer
optimizer = torch.optim.RMSprop(Net.parameters(), lr=1e-3)

# --------------------------------------------
# Logging + GIF setup
# --------------------------------------------
rms_loss = []      # store float loss values
image_list = []

fig_anim, ax_anim = plt.subplots(figsize=(15, 10))

frame_every = 20   # capture a frame every 20 iterations
print_every = 50   # print progress every 50 iterations

# --------------------------------------------
# Training loop
# --------------------------------------------
for it in range(Niter):

    # Forward pass
    y_pred = Net(x)

    # MSE loss
    loss = loss_fn(y_pred, y)

    # Backprop + update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Store loss as float
    rms_loss.append(loss.item())

    # Print occasionally (printing every iter is slow)
    if (it == 0) or ((it + 1) % print_every == 0):
        print(f"Iter: {it + 1:03}/{Niter}  Loss: {loss.item():.5e}")

    # Capture GIF frames occasionally
    if it % frame_every == 0:
        with torch.no_grad():
            y_pred_np = Net(x).detach().cpu().numpy()

        ax_anim.cla()
        ax_anim.set_xlabel("x", fontsize=20)
        ax_anim.set_ylabel("y", fontsize=20)
        ax_anim.set_xlim(-4, 4)
        ax_anim.set_ylim(-1.5, 7.0)

        ax_anim.scatter(x_in, y_in, label="Training Data")
        ax_anim.plot(x_in, y_pred_np, "k--", lw=2, label="NN Prediction")

        ax_anim.text(-3.5, 6.2, f"RMSprop Optimizer (iter={it})", fontsize=14)
        ax_anim.text(-3.5, 5.8, f"Loss = {loss.item():.6f}", fontsize=14)
        ax_anim.legend()

        fig_anim.canvas.draw()
        frame = np.frombuffer(fig_anim.canvas.tostring_rgb(), dtype=np.uint8)
        frame = frame.reshape(fig_anim.canvas.get_width_height()[::-1] + (3,))
        image_list.append(frame)

# --------------------------------------------
# Save GIF
# --------------------------------------------
gif_path = "./RMS_func.gif"
imageio.mimsave(gif_path, image_list, fps=5)
print(f"Saved GIF to: {gif_path}")

# ============================================
# Plot 1: Loss curve (log scale)
# ============================================
fig_loss, ax_loss = plt.subplots(figsize=(12, 7))
ep = np.arange(1, Niter + 1)

ax_loss.semilogy(ep, rms_loss, "m-", lw=2, label="RMSprop Train Loss")
ax_loss.set_xlabel("Iterations")
ax_loss.set_ylabel("MSE Loss (log scale)")
ax_loss.set_title("Training Loss with RMSprop")
ax_loss.legend()
plt.show()

# ============================================
# Plot 2: Final prediction vs true function
# ============================================
with torch.no_grad():
    y_final = Net(x).detach().cpu().numpy()

fig_pred, ax_pred = plt.subplots(figsize=(12, 7))
ax_pred.plot(x_in, y_in, "k-", label="True Function")
ax_pred.scatter(x_in, y_in, label="Training Points")
ax_pred.plot(x_in, y_final, "r-", lw=2, label="Final NN Prediction (RMSprop)")

ax_pred.set_xlabel("x")
ax_pred.set_ylabel("y")
ax_pred.set_title("Final Fit after RMSprop Optimization")
ax_pred.legend()
plt.show()
```

```python
# ============================================
# Adamax Optimization (variant of Adam based on infinity norm)
# + GIF of training progress
# + Loss curve
# + Final prediction plot
# ============================================

import numpy as np
import imageio
import torch
import matplotlib.pyplot as plt

%matplotlib inline
torch.manual_seed(1234)

print("Adamax Optimization")

# --------------------------------------------
# Target function (piecewise)
# --------------------------------------------
def fun_x(x):
    if x < 0.0:
        return 5.0 + np.sin(x) + np.sin(2.0 * x) + np.sin(3.0 * x) + np.sin(4.0 * x)
    else:
        return np.cos(10.0 * x)

# --------------------------------------------
# Build training data
# --------------------------------------------
x_in_l = np.linspace(-np.pi, -1.0e-3, 201)
x_in_r = np.linspace(0.0, np.pi, 501)
x_in = np.concatenate((x_in_l, x_in_r))
y_in = np.array([fun_x(i) for i in x_in])

# Plot training data
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x_in, y_in, 'k-', label="True Function")
ax.plot(x_in, y_in, '.b', label="Training Points")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()

# Convert to tensors
x = torch.from_numpy(x_in.reshape(-1, 1)).float()
y = torch.from_numpy(y_in.reshape(-1, 1)).float()

# --------------------------------------------
# Model + loss + optimizer
# --------------------------------------------
Net = torch.nn.Sequential(
    torch.nn.Linear(1, 100),
    torch.nn.ReLU(),
    torch.nn.Linear(100, 1)
)

loss_fn = torch.nn.MSELoss()
Niter = 500

# Adamax optimizer (-norm variant of Adam)
optimizer = torch.optim.Adamax(Net.parameters(), lr=1e-3)

# --------------------------------------------
# Logging + GIF setup
# --------------------------------------------
adamax_loss = []      # store float losses
image_list = []

fig_anim, ax_anim = plt.subplots(figsize=(15, 10))

frame_every = 20
print_every = 50

# --------------------------------------------
# Training loop
# --------------------------------------------
for it in range(Niter):

    # Forward pass
    y_pred = Net(x)
    loss = loss_fn(y_pred, y)

    # Backprop + update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Store float loss
    adamax_loss.append(loss.item())

    if (it == 0) or ((it + 1) % print_every == 0):
        print(f"Iter: {it + 1:03}/{Niter}  Loss: {loss.item():.5e}")

    # Capture GIF frames occasionally
    if it % frame_every == 0:
        with torch.no_grad():
            y_pred_np = Net(x).detach().cpu().numpy()

        ax_anim.cla()
        ax_anim.set_xlabel("x")
        ax_anim.set_ylabel("y")
        ax_anim.set_xlim(-4, 4)
        ax_anim.set_ylim(-1.5, 7.0)

        ax_anim.scatter(x_in, y_in, label="Training Data")
        ax_anim.plot(x_in, y_pred_np, "k--", lw=2, label="NN Prediction")

        ax_anim.text(-3.5, 6.2, f"Adamax Optimizer (iter={it})", fontsize=14)
        ax_anim.text(-3.5, 5.8, f"Loss = {loss.item():.6f}", fontsize=14)

        ax_anim.legend()

        fig_anim.canvas.draw()
        frame = np.frombuffer(fig_anim.canvas.tostring_rgb(), dtype=np.uint8)
        frame = frame.reshape(fig_anim.canvas.get_width_height()[::-1] + (3,))
        image_list.append(frame)

# --------------------------------------------
# Save GIF
# --------------------------------------------
gif_path = "./Adamax_func.gif"
imageio.mimsave(gif_path, image_list, fps=2)
print(f"Saved GIF to: {gif_path}")

# ============================================
# Plot 1: Loss curve (log scale)
# ============================================
fig_loss, ax_loss = plt.subplots(figsize=(12, 7))
ep = np.arange(1, Niter + 1)

ax_loss.semilogy(ep, adamax_loss, "b-", lw=2, label="Adamax Train Loss")
ax_loss.set_xlabel("Iterations")
ax_loss.set_ylabel("MSE Loss (log scale)")
ax_loss.set_title("Training Loss with Adamax")
ax_loss.legend()
plt.show()

# ============================================
# Plot 2: Final prediction vs true function
# ============================================
with torch.no_grad():
    y_final = Net(x).detach().cpu().numpy()

fig_pred, ax_pred = plt.subplots(figsize=(12, 7))
ax_pred.plot(x_in, y_in, "k-", label="True Function")
ax_pred.scatter(x_in, y_in, label="Training Points")
ax_pred.plot(x_in, y_final, "r-", lw=2, label="Final NN Prediction (Adamax)")

ax_pred.set_xlabel("x")
ax_pred.set_ylabel("y")
ax_pred.set_title("Final Fit after Adamax Optimization")
ax_pred.legend()
plt.show()
```

# Adagrad Optimization (Neural regression) + GIF + Loss plot + Final prediction plot

```python
# ============================================
# Adagrad Optimization (Neural regression) + GIF + Loss plot + Final prediction plot
# ============================================

import numpy as np
import imageio
import torch
import matplotlib.pyplot as plt

%matplotlib inline
torch.manual_seed(1234)

print("Adagrad Optimization")

# --------------------------------------------
# Target function (piecewise)
#   x < 0  : 5 + sum of sines
#   x >= 0 : cos(10x)
# --------------------------------------------
def fun_x(x):
    if x < 0.0:
        return 5.0 + np.sin(x) + np.sin(2.0 * x) + np.sin(3.0 * x) + np.sin(4.0 * x)
    else:
        return np.cos(10.0 * x)

# --------------------------------------------
# Build training data
# --------------------------------------------
x_in_l = np.linspace(-np.pi, -1.0e-3, 201)
x_in_r = np.linspace(0.0, np.pi, 501)
x_in = np.concatenate((x_in_l, x_in_r))
y_in = np.array([fun_x(i) for i in x_in])

# Plot training data
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x_in, y_in, "k-", label="True Function")
ax.plot(x_in, y_in, ".b", label="Training Points")
ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)
ax.legend()
plt.show()

# Convert to tensors (shape [N,1])
x = torch.from_numpy(x_in.reshape(-1, 1)).float()
y = torch.from_numpy(y_in.reshape(-1, 1)).float()

# --------------------------------------------
# Model + loss + optimizer
# --------------------------------------------
Net = torch.nn.Sequential(
    torch.nn.Linear(1, 100),
    torch.nn.ReLU(),
    torch.nn.Linear(100, 1),
)

loss_fn = torch.nn.MSELoss()
Niter = 500

# Adagrad optimizer (accumulates squared gradients; LR effectively decays over time)
optimizer = torch.optim.Adagrad(Net.parameters(), lr=1e-3)

# --------------------------------------------
# Logging + GIF setup
# --------------------------------------------
adagrad_loss = []   # store float losses
image_list = []

fig_anim, ax_anim = plt.subplots(figsize=(15, 10))

frame_every = 20    # capture a frame every 20 iterations
print_every = 50    # print every 50 iterations

# --------------------------------------------
# Training loop
# --------------------------------------------
for it in range(Niter):
    # Forward pass
    y_pred = Net(x)

    # Compute loss
    loss = loss_fn(y_pred, y)

    # Backprop + update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Store float loss
    adagrad_loss.append(loss.item())

    # Print occasionally (printing every iter is slow)
    if (it == 0) or ((it + 1) % print_every == 0):
        print(f"Iter: {it + 1:03}/{Niter}  Loss: {loss.item():.5e}")

    # Capture GIF frame occasionally
    if it % frame_every == 0:
        with torch.no_grad():
            y_pred_np = Net(x).detach().cpu().numpy()

        ax_anim.cla()
        ax_anim.set_xlabel("x", fontsize=20)
        ax_anim.set_ylabel("y", fontsize=20)
        ax_anim.set_xlim(-4, 4)
        ax_anim.set_ylim(-1.5, 7.0)

        ax_anim.scatter(x_in, y_in, label="Training Data")
        ax_anim.plot(x_in, y_pred_np, "k--", lw=2, label="NN Prediction")

        ax_anim.text(-3.5, 6.2, f"Adagrad Optimizer (iter={it})", fontsize=14)
        ax_anim.text(-3.5, 5.8, f"Loss = {loss.item():.6f}", fontsize=14)
        ax_anim.legend()

        fig_anim.canvas.draw()
        frame = np.frombuffer(fig_anim.canvas.tostring_rgb(), dtype=np.uint8)
        frame = frame.reshape(fig_anim.canvas.get_width_height()[::-1] + (3,))
        image_list.append(frame)

# --------------------------------------------
# Save GIF
# --------------------------------------------
gif_path = "./Adagrad_func.gif"
imageio.mimsave(gif_path, image_list, fps=5)
print(f"Saved GIF to: {gif_path}")

# ============================================
# Plot 1: Loss curve (log scale)
# ============================================
fig_loss, ax_loss = plt.subplots(figsize=(12, 7))
ep = np.arange(1, Niter + 1)

ax_loss.semilogy(ep, adagrad_loss, "g-", lw=2, label="Adagrad Train Loss")
ax_loss.set_xlabel("Iterations")
ax_loss.set_ylabel("MSE Loss (log scale)")
ax_loss.set_title("Training Loss with Adagrad")
ax_loss.legend()
plt.show()

# ============================================
# Plot 2: Final prediction vs true function
# ============================================
with torch.no_grad():
    y_final = Net(x).detach().cpu().numpy()

fig_pred, ax_pred = plt.subplots(figsize=(12, 7))
ax_pred.plot(x_in, y_in, "k-", label="True Function")
ax_pred.scatter(x_in, y_in, label="Training Points")
ax_pred.plot(x_in, y_final, "r-", lw=2, label="Final NN Prediction (Adagrad)")

ax_pred.set_xlabel("x")
ax_pred.set_ylabel("y")
ax_pred.set_title("Final Fit after Adagrad Optimization")
ax_pred.legend()
plt.show()
```

# Adadelta Optimization (Neural regression) + GIF + Loss plot + Final prediction plot

# Combination of Adam + L-BFGS


```python
# ============================================
# Combination of Adam + L-BFGS (two-phase training)
# 1) Warm-start with Adam
# 2) Fine-tune with L-BFGS
#
# Outputs:
# - GIF showing prediction improving over time
# - Loss plot (Adam vs L-BFGS)
# - Final prediction plot vs the true function
# ============================================

import numpy as np
import imageio
import torch
import matplotlib.pyplot as plt

%matplotlib inline
torch.manual_seed(1234)

# --------------------------------------------
# Target function (piecewise)
# --------------------------------------------
def fun_x(x):
    if x < 0.0:
        return 5.0 + np.sin(x) + np.sin(2.0 * x) + np.sin(3.0 * x) + np.sin(4.0 * x)
    else:
        return np.cos(10.0 * x)

# --------------------------------------------
# Build training data
# --------------------------------------------
x_in_l = np.linspace(-np.pi, -1.0e-3, 201)
x_in_r = np.linspace(0.0, np.pi, 501)
x_in = np.concatenate((x_in_l, x_in_r), axis=0)
y_in = np.array([fun_x(i) for i in x_in])

# Plot the true function once
fig_true, ax_true = plt.subplots(figsize=(10, 6))
ax_true.plot(x_in, y_in, "k-", label="True Function")
ax_true.set_xlabel("x")
ax_true.set_ylabel("y")
ax_true.legend()
plt.show()

# Convert to torch tensors
x = torch.from_numpy(x_in.reshape(-1, 1)).float()
y = torch.from_numpy(y_in.reshape(-1, 1)).float()

# --------------------------------------------
# Model: 1 -> 100 (ReLU) -> 1
# --------------------------------------------
Net = torch.nn.Sequential(
    torch.nn.Linear(1, 100),
    torch.nn.ReLU(),
    torch.nn.Linear(100, 1)
)

loss_fn = torch.nn.MSELoss()

# --------------------------------------------
# Training settings
# --------------------------------------------
Niter_adam = 30
Niter_lbfgs_steps = 50   # number of outer "epochs" calling LBFGS.step(...)

optimizer_adam = torch.optim.Adam(Net.parameters(), lr=1e-3)

# IMPORTANT:
# In PyTorch, LBFGS has an internal loop controlled by max_iter
# max_iter should be moderate (e.g., 20 or 50), otherwise each .step can be extremely expensive.
optimizer_lbfgs = torch.optim.LBFGS(
    Net.parameters(),
    history_size=8,
    max_iter=50,         # FIX: your original 500000 would be very slow
    line_search_fn="strong_wolfe"
)

# --------------------------------------------
# Logging (store floats)
# --------------------------------------------
adam_loss = []
lbfgs_loss = []

# --------------------------------------------
# GIF setup
# --------------------------------------------
fig_anim, ax_anim = plt.subplots(figsize=(15, 10))
image_list = []
frame_every_adam = 4
frame_every_lbfgs = 2

def capture_frame(title_text, loss_value):
    """Render current prediction and capture an RGB frame for GIF."""
    ax_anim.cla()

    with torch.no_grad():
        y_pred_np = Net(x).detach().cpu().numpy()

    ax_anim.set_xlabel("x", fontsize=20)
    ax_anim.set_ylabel("y", fontsize=20)
    ax_anim.set_xlim(-4, 4)
    ax_anim.set_ylim(-1.5, 7.0)

    ax_anim.scatter(x_in, y_in, label="Training Data")
    ax_anim.plot(x_in, y_pred_np, "k--", lw=2, label="NN Prediction")

    ax_anim.text(-3.5, 6.2, title_text, fontsize=16)
    ax_anim.text(-3.5, 5.7, f"Loss = {loss_value:.6f}", fontsize=16)
    ax_anim.legend()

    fig_anim.canvas.draw()
    frame = np.frombuffer(fig_anim.canvas.tostring_rgb(), dtype=np.uint8)
    frame = frame.reshape(fig_anim.canvas.get_width_height()[::-1] + (3,))
    image_list.append(frame)

# ============================================
# Phase 1: Adam warm-start
# ============================================
print("Phase 1: Adam Optimization")
print_every = 10

for it in range(Niter_adam):
    y_pred = Net(x)
    loss = loss_fn(y_pred, y)

    optimizer_adam.zero_grad()
    loss.backward()
    optimizer_adam.step()

    adam_loss.append(loss.item())

    if (it == 0) or ((it + 1) % print_every == 0):
        print(f"Adam Iter: {it + 1:02}/{Niter_adam}  Loss: {loss.item():.5e}")

    if it % frame_every_adam == 0:
        capture_frame(title_text=f"Adam Optimizer (iter={it})", loss_value=loss.item())

# ============================================
# Phase 2: L-BFGS fine-tune
# ============================================
print("\nPhase 2: L-BFGS Optimization")

def closure():
    """LBFGS requires a closure that reevaluates the model and returns the loss."""
    optimizer_lbfgs.zero_grad()
    y_pred = Net(x)
    loss = loss_fn(y_pred, y)
    loss.backward()
    return loss

for it in range(Niter_lbfgs_steps):
    # optimizer.step runs its internal LBFGS iterations (up to max_iter)
    loss = optimizer_lbfgs.step(closure)

    # loss returned may be a tensor or float depending on version; normalize to float
    loss_val = float(loss) if not torch.is_tensor(loss) else loss.item()
    lbfgs_loss.append(loss_val)

    if (it == 0) or ((it + 1) % 5 == 0):
        print(f"LBFGS Step: {it + 1:02}/{Niter_lbfgs_steps}  Loss: {loss_val:.5e}")

    if it % frame_every_lbfgs == 0:
        capture_frame(title_text=f"L-BFGS Optimizer (step={it})", loss_value=loss_val)

# --------------------------------------------
# Save GIF
# --------------------------------------------
gif_path = "./Adam_and_LBFGS.gif"
imageio.mimsave(gif_path, image_list, fps=1)
print(f"\nSaved GIF to: {gif_path}")

# ============================================
# Plot 1: Loss curves (Adam vs L-BFGS)
# ============================================
fig_loss, ax_loss = plt.subplots(figsize=(12, 7))

ax_loss.semilogy(np.arange(len(adam_loss)), adam_loss, label="Adam Loss")
ax_loss.semilogy(np.arange(len(lbfgs_loss)), lbfgs_loss, label="L-BFGS Loss")

ax_loss.set_xlabel("Iterations / Steps")
ax_loss.set_ylabel("MSE Loss (log scale)")
ax_loss.set_title("Loss Curves: Adam (warm-start) + L-BFGS (fine-tune)")
ax_loss.legend()
plt.show()

# ============================================
# Plot 2: Final prediction vs true function
# ============================================
with torch.no_grad():
    y_final = Net(x).detach().cpu().numpy()

fig_pred, ax_pred = plt.subplots(figsize=(12, 7))
ax_pred.plot(x_in, y_in, "k-", label="True Function")
ax_pred.scatter(x_in, y_in, label="Training Points")
ax_pred.plot(x_in, y_final, "r-", lw=2, label="Final NN Prediction (Adam + L-BFGS)")

ax_pred.set_xlabel("x")
ax_pred.set_ylabel("y")
ax_pred.set_title("Final Fit after Adam + L-BFGS")
ax_pred.legend()
plt.show()
```
