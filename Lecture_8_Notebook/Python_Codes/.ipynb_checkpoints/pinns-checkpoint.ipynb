## PINN for Burger's Equation in TensorFlow

```python
import sys
sys.path.insert(0, 'Utilities/')
import os

from scipy.interpolate import griddata
from pyDOE import lhs
from plotting import newfig, savefig
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time
import scipy.io
```

```python
# ---------------------------------------------------------
# Reproducibility: fix random seeds so results are repeatable
# ---------------------------------------------------------
np.random.seed(seed=1234)          # NumPy random seed
tf.random.set_seed(1234)           # TensorFlow random seed

# Disable TensorFloat-32 for more controlled numerical behavior
tf.config.experimental.enable_tensor_float_32_execution(False)

# Optional mixed precision setting (currently disabled)
# os.environ['TF_ENABLE_AUTO_MIXED_PRECISION'] = '1'


# ---------------------------------------------------------
# Weight initialization function
# ---------------------------------------------------------
def hyper_initial(size):
    """
    Initialize weights for one layer of the neural network.

    Parameters:
        size : list or tuple
            [input_dimension, output_dimension]

    Returns:
        tf.Variable
            Weight matrix initialized using truncated normal distribution
            with Xavier/He-like scaling.
    """
    in_dim = size[0]
    out_dim = size[1]

    # Standard deviation scaled with layer size
    std = np.sqrt(2.0 / (in_dim + out_dim))

    # Create trainable weight matrix
    return tf.Variable(tf.random.truncated_normal(shape=size, stddev=std))


# ---------------------------------------------------------
# Deep Neural Network forward pass
# ---------------------------------------------------------
def DNN(X, W, b):
    """
    Forward propagation through the neural network.

    Parameters:
        X : input tensor
        W : list of weight matrices
        b : list of bias vectors

    Returns:
        Y : network output
    """
    A = X                     # Start with input
    L = len(W)                # Number of layers with weights

    # Hidden layers: linear transform + tanh activation
    for i in range(L - 1):
        A = tf.tanh(tf.add(tf.matmul(A, W[i]), b[i]))

    # Output layer: linear activation
    Y = tf.add(tf.matmul(A, W[-1]), b[-1])
    return Y


# ---------------------------------------------------------
# Helper function: collect trainable variables
# ---------------------------------------------------------
def train_vars(W, b):
    """
    Return all trainable parameters (weights + biases)
    as one list for the optimizer.
    """
    return W + b


# ---------------------------------------------------------
# Neural network solution u(x,t)
# ---------------------------------------------------------
def net_u(x, t, w, b):
    """
    Predict u(x,t) using the neural network.

    The inputs x and t are concatenated to form [x, t].
    """
    u = DNN(tf.concat([x, t], 1), w, b)
    return u


# ---------------------------------------------------------
# Physics residual for Burgers' equation
# f = u_t + u*u_x - nu*u_xx
# ---------------------------------------------------------
@tf.function()
def net_f(x, t, W, b, nu):
    """
    Compute the PDE residual using automatic differentiation.

    Burgers' equation:
        u_t + u*u_x - nu*u_xx = 0

    If the network satisfies the PDE perfectly, then f should be 0.
    """
    with tf.GradientTape(persistent=True) as tape1:
        tape1.watch([x, t])

        with tf.GradientTape(persistent=True) as tape2:
            tape2.watch([x, t])

            # Network prediction
            u = net_u(x, t, W, b)

        # First derivatives
        u_t = tape2.gradient(u, t)
        u_x = tape2.gradient(u, x)

    # Second derivative with respect to x
    u_xx = tape1.gradient(u_x, x)

    # PDE residual
    f = u_t + u * u_x - nu * u_xx
    return f


# ---------------------------------------------------------
# One training step
# ---------------------------------------------------------
@tf.function()
def train_step(W, b, X_u_train_tf, u_train_tf, X_f_train_tf, opt, nu):
    """
    Perform one optimization step.

    Loss = \lambda_d * data loss + \lambda_f*physics loss
         = MSE on known solution points + MSE on PDE residual points
    """

    # Known data points (initial/boundary condition points)
    x_u = X_u_train_tf[:, 0:1]
    t_u = X_u_train_tf[:, 1:2]

    # Collocation points where PDE residual is enforced
    x_f = X_f_train_tf[:, 0:1]
    t_f = X_f_train_tf[:, 1:2]

    with tf.GradientTape() as tape:
        tape.watch([W, b])

        # Network prediction at data points
        u_nn = net_u(x_u, t_u, W, b)

        # PDE residual at collocation points
        f_nn = net_f(x_f, t_f, W, b, nu)

        # Total loss:
        # 1. Fit the known training data
        # 2. Enforce the PDE
        loss = tf.reduce_mean(tf.square(u_nn - u_train_tf)) + \
               tf.reduce_mean(tf.square(f_nn))

    # Compute gradients of loss with respect to weights and biases
    grads = tape.gradient(loss, train_vars(W, b))

    # Update parameters
    opt.apply_gradients(zip(grads, train_vars(W, b)))

    return loss


# ---------------------------------------------------------
# Problem setup
# ---------------------------------------------------------
nu = 0.01 / np.pi   # Viscosity coefficient in Burgers' equation
N_u = 100           # Number of initial/boundary training points
N_f = 10000         # Number of collocation points for PDE residual
Nmax = 20000        # Maximum number of training iterations


# ---------------------------------------------------------
# Network architecture
# layers = [input, hidden1, hidden2, ..., output]
# ---------------------------------------------------------
layers = [2, 20, 20, 20, 20, 20, 20, 20, 20, 1]

L = len(layers)

# Initialize weights and biases for each layer
W = [hyper_initial([layers[l - 1], layers[l]]) for l in range(1, L)]
b = [tf.Variable(tf.zeros([1, layers[l]])) for l in range(1, L)]


# ---------------------------------------------------------
# Load Burgers' equation dataset
# ---------------------------------------------------------
data = scipy.io.loadmat('./Data/burgers_shock.mat')

t = data['t'].flatten()[:, None]             # time values
x = data['x'].flatten()[:, None]             # space values
Exact = np.real(data['usol']).T              # exact solution u(x,t)

# Create full mesh of (x,t) points
X, T = np.meshgrid(x, t)

# Flatten the mesh into a list of coordinate pairs
X_star = np.hstack((X.flatten()[:, None], T.flatten()[:, None]))
u_star = Exact.flatten()[:, None]


# ---------------------------------------------------------
# Domain bounds
# ---------------------------------------------------------
lb = X_star.min(0)   # lower bound [x_min, t_min]
ub = X_star.max(0)   # upper bound [x_max, t_max]


# ---------------------------------------------------------
# Construct training data from initial and boundary conditions
# ---------------------------------------------------------

# Initial condition: t = 0
xx1 = np.hstack((X[0:1, :].T, T[0:1, :].T))
uu1 = Exact[0:1, :].T

# Boundary condition at x = -1
xx2 = np.hstack((X[:, 0:1], T[:, 0:1]))
uu2 = Exact[:, 0:1]

# Boundary condition at x = 1
xx3 = np.hstack((X[:, -1:], T[:, -1:]))
uu3 = Exact[:, -1:]


# Stack all known data points together
X_u_train = np.vstack([xx1, xx2, xx3])
u_train = np.vstack([uu1, uu2, uu3])


# ---------------------------------------------------------
# Generate collocation points inside the domain using Latin Hypercube Sampling
# ---------------------------------------------------------
X_f_train = lb + (ub - lb) * lhs(2, N_f)

# Also include boundary/initial points in collocation set
X_f_train = np.vstack((X_f_train, X_u_train))


# ---------------------------------------------------------
# Randomly choose N_u training points from available known points
# ---------------------------------------------------------
idx = np.random.choice(X_u_train.shape[0], N_u, replace=False)

X_u_train = X_u_train[idx, :]
u_train = u_train[idx, :]


# ---------------------------------------------------------
# Convert NumPy arrays to TensorFlow tensors
# ---------------------------------------------------------
X_u_train_tf = tf.convert_to_tensor(X_u_train, dtype=tf.float32)
u_train_tf = tf.convert_to_tensor(u_train, dtype=tf.float32)
X_f_train_tf = tf.convert_to_tensor(X_f_train, dtype=tf.float32)


# ---------------------------------------------------------
# Optimizer setup
# ---------------------------------------------------------
lr = 1e-3
optimizer = tf.optimizers.Adam(learning_rate=lr)


# ---------------------------------------------------------
# Training loop
# ---------------------------------------------------------
start_time = time.time()
n = 0
loss = []

while n <= Nmax:
    # Perform one training step
    loss_ = train_step(W, b, X_u_train_tf, u_train_tf, X_f_train_tf, optimizer, nu)
    loss.append(loss_)

    # Print progress every 1000 iterations
    if n % 1000 == 0:
        print(f"Iteration is: {n} and loss is: {loss_}")

    n += 1


# ---------------------------------------------------------
# Training finished
# ---------------------------------------------------------
elapsed = time.time() - start_time
print('Training time: %.4f' % elapsed)
```

```python
# ---------------------------------------------------------
# Prediction function
# ---------------------------------------------------------
def predict(X_star_tf, w, b):
    """
    Predict the solution u(x,t) at all given test points.

    Parameters:
        X_star_tf : tensor of shape [N, 2]
            Each row contains one point [x, t]
        w : list of weights
        b : list of biases

    Returns:
        u_pred : predicted solution values at all test points
    """

    # Extract x and t from the input tensor
    x_star = X_star_tf[:, 0:1]
    t_star = X_star_tf[:, 1:2]

    # Pass x and t through the trained neural network
    u_pred = net_u(x_star, t_star, w, b)

    return u_pred


# ---------------------------------------------------------
# Convert all test coordinates into TensorFlow tensor
# ---------------------------------------------------------
X_star_tf = tf.convert_to_tensor(X_star, dtype=tf.float32)

# Predict the solution at all test points
u_pred = predict(X_star_tf, W, b)


# ---------------------------------------------------------
# Compute relative L2 error between exact and predicted solution
# ---------------------------------------------------------
error_u = np.linalg.norm(u_star - u_pred, 2) / np.linalg.norm(u_star, 2)

# Print the prediction error
print('Error u: %e' % (error_u))


# ---------------------------------------------------------
# Interpolate the predicted values back onto the mesh grid
# ---------------------------------------------------------
U_pred = griddata(
    X_star,                       # known coordinate points
    u_pred.numpy().flatten(),     # predicted values at those points
    (X, T),                       # target mesh grid
    method='cubic'                # interpolation method
)

# Compute percentage error over the full field
Error = 100 * np.linalg.norm(Exact - U_pred) / np.linalg.norm(U_pred)


# ---------------------------------------------------------
# Create figure for visualization
# ---------------------------------------------------------
fig, ax = newfig(1.0, 1.1)
ax.axis('off')   # Turn off default axes for the outer figure


# =========================================================
# Row 0: Heatmap of predicted solution u(t,x)
# =========================================================
gs0 = gridspec.GridSpec(1, 2)
gs0.update(top=1 - 0.06, bottom=1 - 1/3, left=0.15, right=0.85, wspace=0)

ax = plt.subplot(gs0[:, :])

# Plot the 2D predicted solution as an image
h = ax.imshow(
    U_pred.T,
    interpolation='nearest',
    cmap='rainbow',
    extent=[t.min(), t.max(), x.min(), x.max()],
    origin='lower',
    aspect='auto'
)

# Add colorbar to show the value of u(t,x)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(h, cax=cax)

# Plot training data points (initial and boundary points) on top of the heatmap
ax.plot(
    X_u_train[:, 1], X_u_train[:, 0],
    'kx',
    label='Data (%d points)' % (u_train.shape[0]),
    markersize=4,
    clip_on=False
)

# Optional: plot some residual points if needed
# ax.plot(X_f_train[0:100,1], X_f_train[0:100,0], 'bx',
#         label='Residual points', markersize=4, clip_on=False)

# Draw vertical white lines at selected time snapshots
line = np.linspace(x.min(), x.max(), 2)[:, None]
ax.plot(t[25] * np.ones((2, 1)), line, 'w-', linewidth=1)
ax.plot(t[50] * np.ones((2, 1)), line, 'w-', linewidth=1)
ax.plot(t[75] * np.ones((2, 1)), line, 'w-', linewidth=1)

# Label axes and add title
ax.set_xlabel('$t$')
ax.set_ylabel('$x$')
ax.legend(frameon=False, loc='best')
ax.set_title('$u(t,x)$', fontsize=10)


# =========================================================
# Row 1: 1D slices comparing exact vs predicted solution
# =========================================================
gs1 = gridspec.GridSpec(1, 3)
gs1.update(top=1 - 1/3, bottom=0, left=0.1, right=0.9, wspace=0.5)


# ---------------------------------------------------------
# Slice at t = 0.25
# ---------------------------------------------------------
ax = plt.subplot(gs1[0, 0])

ax.plot(x, Exact[25, :], 'b-', linewidth=2, label='Exact')
ax.plot(x, U_pred[25, :], 'r--', linewidth=2, label='Prediction')

ax.set_xlabel('$x$')
ax.set_ylabel('$u(t,x)$')
ax.set_title('$t = 0.25$', fontsize=10)

# Make axes square and fix display range
ax.axis('square')
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])


# ---------------------------------------------------------
# Slice at t = 0.50
# ---------------------------------------------------------
ax = plt.subplot(gs1[0, 1])

ax.plot(x, Exact[50, :], 'b-', linewidth=2, label='Exact')
ax.plot(x, U_pred[50, :], 'r--', linewidth=2, label='Prediction')

ax.set_xlabel('$x$')
ax.set_ylabel('$u(t,x)$')
ax.set_title('$t = 0.50$', fontsize=10)

ax.axis('square')
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])

# Add shared legend below this middle plot
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.35), ncol=5, frameon=False)


# ---------------------------------------------------------
# Slice at t = 0.75
# ---------------------------------------------------------
ax = plt.subplot(gs1[0, 2])

ax.plot(x, Exact[75, :], 'b-', linewidth=2, label='Exact')
ax.plot(x, U_pred[75, :], 'r--', linewidth=2, label='Prediction')

ax.set_xlabel('$x$')
ax.set_ylabel('$u(t,x)$')
ax.set_title('$t = 0.75$', fontsize=10)

ax.axis('square')
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])
```

## PINN for Burger's Equation in JAX

```python
import jax.numpy as jnp
from jax import grad, jit, vmap
from jax import random
import numpy as np
import scipy.io
from pyDOE import lhs
from jax.nn import tanh, relu
import sys
from jax.example_libraries import optimizers
from tqdm import trange
sys.path.insert(0, 'Utilities/')
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from plotting import newfig, savefig
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# ---------------------------------------------------------
# Initialize parameters for one neural network layer
# ---------------------------------------------------------
def random_layer_params(m, n, key, scale):
    """
    Initialize weights and biases for one layer.

    Parameters:
        m : int
            Number of input neurons
        n : int
            Number of output neurons
        key : JAX random key
            Random key for parameter generation
        scale : float
            Scaling factor for weight initialization

    Returns:
        (W, b) : tuple
            W = weight matrix of shape (m, n)
            b = bias vector of shape (n,)
    """
    w_key, b_key = random.split(key)

    # Randomly initialize weights and set biases to zero
    return scale * random.normal(w_key, (m, n)), jnp.zeros(n)


# ---------------------------------------------------------
# Initialize all network parameters
# ---------------------------------------------------------
def init_network_params(sizes, key):
    """
    Initialize the full neural network.

    Parameters:
        sizes : list
            Example: [2, 20, 20, 1]
            means input layer = 2, two hidden layers of 20, output = 1
        key : JAX random key

    Returns:
        params : list of (W, b) tuples for all layers
    """
    keys = random.split(key, len(sizes))

    return [
        random_layer_params(m, n, k, 2.0 / (jnp.sqrt(m + n)))
        for m, n, k in zip(sizes[:-1], sizes[1:], keys)
    ]


# ---------------------------------------------------------
# Forward pass through the neural network
# ---------------------------------------------------------
@jit
def predict(params, X, lb, ub):
    """
    Compute the neural network output for input X.

    Parameters:
        params : network parameters
        X : input data
        lb : lower bound of domain
        ub : upper bound of domain

    Returns:
        H : predicted output
    """

    # Normalize inputs to the range [-1, 1]
    H = 2.0 * (X - lb) / (ub - lb) - 1.0

    # Pass through all hidden layers with tanh activation
    for w, b in params[:-1]:
        H = tanh(jnp.dot(H, w) + b)

    # Final layer is linear
    final_w, final_b = params[-1]
    H = jnp.dot(H, final_w) + final_b

    return H


# ---------------------------------------------------------
# Predict u(x,t)
# ---------------------------------------------------------
@jit
def net_u(params, x, t, lb, ub):
    """
    Predict the solution u(x,t) for one point.
    """
    x_con = jnp.array([x, t])     # combine x and t into one input vector
    y_pred = predict(params, x_con, lb, ub)
    return y_pred


# ---------------------------------------------------------
# Helper function for gradient calculation
# ---------------------------------------------------------
@jit
def net_u_grad(params, x, t, lb, ub):
    """
    Return scalar output u(x,t), used for differentiation.
    """
    x_con = jnp.array([x, t])
    y_pred = predict(params, x_con, lb, ub)
    return y_pred[0]


# ---------------------------------------------------------
# First derivatives u_t and u_x
# ---------------------------------------------------------
def net_f(params, lb, ub):
    """
    Return JIT-compiled functions for first derivatives:
        u_t = du/dt
        u_x = du/dx
    """

    def u_t(x, t):
        ut = grad(net_u_grad, argnums=2)(params, x, t, lb, ub)
        return ut

    def u_x(x, t):
        ux = grad(net_u_grad, argnums=1)(params, x, t, lb, ub)
        return ux

    return jit(u_t), jit(u_x)


# ---------------------------------------------------------
# Second derivative u_xx
# ---------------------------------------------------------
def net_fxx(params, lb, ub):
    """
    Return JIT-compiled function for second derivative:
        u_xx = d²u/dx²
    """

    def u_xx(x, t):
        _, u_x = net_f(params, lb, ub)
        ux = grad(u_x, argnums=0)(x, t)
        return ux

    return jit(u_xx)


# ---------------------------------------------------------
# Data loss: fit initial and boundary data
# ---------------------------------------------------------
@jit
def loss_data(params, x, t, lb, ub, u_train):
    """
    Compute mean squared error between prediction and known data.
    """

    # Apply net_u to all (x,t) points using vectorization
    u_pred = vmap(net_u, (None, 0, 0, None, None))(params, x, t, lb, ub)

    # Mean squared error
    loss = jnp.mean((u_pred - u_train) ** 2)
    return loss


# ---------------------------------------------------------
# Physics loss: enforce Burgers' equation
# ---------------------------------------------------------
@jit
def loss_f(params, x, t, lb, ub, nu):
    """
    Compute the PDE residual loss for Burgers' equation:

        u_t + u*u_x - nu*u_xx = 0
    """

    # Predict u at collocation points
    u = vmap(net_u, (None, 0, 0, None, None))(params, x, t, lb, ub)

    # Get derivative functions
    u_tf, u_xf = net_f(params, lb, ub)
    u_xxf = net_fxx(params, lb, ub)

    # Evaluate derivatives at all collocation points
    u_t = vmap(u_tf, (0, 0))(x, t)
    u_x = vmap(u_xf, (0, 0))(x, t)
    u_xx = vmap(u_xxf, (0, 0))(x, t)

    # Burgers' equation residual
    res = u_t + u.flatten() * u_x - nu * u_xx

    # Mean squared residual
    loss_f = jnp.mean((res.flatten()) ** 2)
    return loss_f


# ---------------------------------------------------------
# Predict u for many test points
# ---------------------------------------------------------
@jit
def predict_u(params, x_star, t_star, lb, ub):
    """
    Predict u(x,t) for a full set of test points.
    """
    u_pred = vmap(net_u, (None, 0, 0, None, None))(
        params, x_star, t_star, lb, ub
    )
    return u_pred


# ---------------------------------------------------------
# Total loss = data loss + physics loss
# ---------------------------------------------------------
def loss_fn(params, x_f, t_f, x_d, t_d, lb, ub, nu, y_d):
    """
    Combine data loss and residual loss.
    """
    loss_res = loss_f(params, x_f, t_f, lb, ub, nu)
    data_loss = loss_data(params, x_d, t_d, lb, ub, y_d)
    return loss_res + data_loss


# ---------------------------------------------------------
# One optimizer step
# ---------------------------------------------------------
@jit
def step(istep, opt_state, t_d, x_d, y_d, t_f, x_f, lb, ub):
    """
    Perform one optimization step using Adam.
    """
    param = get_params(opt_state)

    # Gradient of total loss with respect to parameters
    g = grad(loss_fn, argnums=0)(param, x_f, t_f, x_d, t_d, lb, ub, nu, y_d)

    # Update optimizer state
    return opt_update(istep, g, opt_state)


# ---------------------------------------------------------
# Main training script
# ---------------------------------------------------------
if __name__ == "__main__":

    # Physical parameter: viscosity
    nu = 0.01 / np.pi

    # Number of training points from data
    N_u = 100

    # Number of collocation points for PDE residual
    N_f = 10000

    # Network architecture
    layers = [2, 20, 20, 20, 20, 20, 20, 20, 20, 20, 1]

    # -----------------------------------------------------
    # Load Burgers' equation dataset
    # -----------------------------------------------------
    data = scipy.io.loadmat('Data/burgers_shock.mat')

    t = data['t'].flatten()[:, None]
    x = data['x'].flatten()[:, None]
    Exact = np.real(data['usol']).T

    # Create meshgrid for full domain
    X, T = np.meshgrid(x, t)

    # Flatten coordinates into test points
    X_star = np.hstack((X.flatten()[:, None], T.flatten()[:, None]))
    u_star = Exact.flatten()[:, None]

    # Domain bounds
    lb = X_star.min(0)
    ub = X_star.max(0)

    # -----------------------------------------------------
    # Initial condition: t = 0
    # -----------------------------------------------------
    xx1 = np.hstack((X[0:1, :].T, T[0:1, :].T))
    uu1 = Exact[0:1, :].T

    # -----------------------------------------------------
    # Boundary condition at x = -1
    # -----------------------------------------------------
    xx2 = np.hstack((X[:, 0:1], T[:, 0:1]))
    uu2 = Exact[:, 0:1]

    # -----------------------------------------------------
    # Boundary condition at x = 1
    # -----------------------------------------------------
    xx3 = np.hstack((X[:, -1:], T[:, -1:]))
    uu3 = Exact[:, -1:]

    # Stack all known points together
    X_u_train = np.vstack([xx1, xx2, xx3])

    # Generate collocation points using Latin Hypercube Sampling
    X_f_train = lb + (ub - lb) * lhs(2, N_f)

    # Add known data points into collocation set
    X_f_train = np.vstack((X_f_train, X_u_train))

    # Stack all exact values for known points
    u_train = np.vstack([uu1, uu2, uu3])

    # Randomly select N_u training points from known data
    idx = np.random.choice(X_u_train.shape[0], N_u, replace=False)
    X_u_train = X_u_train[idx, :]
    u_train = u_train[idx, :]

    # Separate x and t for data points
    x_d = X_u_train[:, 0]
    t_d = X_u_train[:, 1]

    # Separate x and t for collocation points
    x_f = X_f_train[:, 0]
    t_f = X_f_train[:, 1]

    # Separate x and t for full test grid
    x_star = X_star[:, 0]
    t_star = X_star[:, 1]

    # -----------------------------------------------------
    # Initialize network parameters
    # -----------------------------------------------------
    params = init_network_params(layers, random.PRNGKey(1234))

    # Adam optimizer
    opt_init, opt_update, get_params = optimizers.adam(5e-4)
    opt_state = opt_init(params)

    # Number of iterations
    nIter = 200 + 1

    # Lists to store losses
    ld_list = []
    lf_list = []

    # Progress bar
    pbar = trange(nIter)

    # -----------------------------------------------------
    # Training loop
    # -----------------------------------------------------
    for it in pbar:
        opt_state = step(it, opt_state, t_d, x_d, u_train, t_f, x_f, lb, ub)

        # Monitor losses during training
        if it % 1 == 0:
            params = get_params(opt_state)

            l_d = loss_data(params, x_d, t_d, lb, ub, u_train)
            l_f = loss_f(params, x_f, t_f, lb, ub, nu)

            pbar.set_postfix({'Loss': l_d, 'loss_physics': l_f})

            ld_list.append(l_d)
            lf_list.append(l_f)

    # -----------------------------------------------------
    # Prediction after training
    # -----------------------------------------------------
    u_pred = predict_u(params, x_star, t_star, lb, ub)

    # Relative L2 error
    error_u = jnp.linalg.norm(u_star - u_pred, 2) / np.linalg.norm(u_star, 2)
    print('Error u: %e' % (error_u))

    # Save loss histories
    np.save("ld_list.npy", np.array(ld_list), allow_pickle=True)
    np.save("lf_list.npy", np.array(lf_list), allow_pickle=True)

    # -----------------------------------------------------
    # Interpolate prediction back to mesh for plotting
    # -----------------------------------------------------
    U_pred = griddata(X_star, u_pred.flatten(), (X, T), method='cubic')

    # Pointwise absolute error
    Error = np.abs(Exact - U_pred)

    # -----------------------------------------------------
    # Plotting section
    # -----------------------------------------------------
    fig, ax = newfig(1.0, 1.1)
    ax.axis('off')

    # =====================================================
    # Row 0: Heatmap of predicted solution u(t,x)
    # =====================================================
    gs0 = gridspec.GridSpec(1, 2)
    gs0.update(top=1 - 0.06, bottom=1 - 1/3, left=0.15, right=0.85, wspace=0)
    ax = plt.subplot(gs0[:, :])

    h = ax.imshow(
        U_pred.T,
        interpolation='nearest',
        cmap='rainbow',
        extent=[t.min(), t.max(), x.min(), x.max()],
        origin='lower',
        aspect='auto'
    )

    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    fig.colorbar(h, cax=cax)

    # Plot training data locations
    ax.plot(
        X_u_train[:, 1], X_u_train[:, 0],
        'kx',
        label='Data (%d points)' % (u_train.shape[0]),
        markersize=4,
        clip_on=False
    )

    # Draw white vertical lines showing where slices are taken
    line = np.linspace(x.min(), x.max(), 2)[:, None]
    ax.plot(t[25] * np.ones((2, 1)), line, 'w-', linewidth=1)
    ax.plot(t[50] * np.ones((2, 1)), line, 'w-', linewidth=1)
    ax.plot(t[75] * np.ones((2, 1)), line, 'w-', linewidth=1)

    ax.set_xlabel('$t$')
    ax.set_ylabel('$x$')
    ax.legend(frameon=False, loc='best')
    ax.set_title('$u(t,x)$', fontsize=10)

    # =====================================================
    # Row 1: Compare exact vs predicted slices
    # =====================================================
    gs1 = gridspec.GridSpec(1, 3)
    gs1.update(top=1 - 1/3, bottom=0, left=0.1, right=0.9, wspace=0.5)

    # Slice at t = 0.25
    ax = plt.subplot(gs1[0, 0])
    ax.plot(x, Exact[25, :], 'b-', linewidth=2, label='Exact')
    ax.plot(x, U_pred[25, :], 'r--', linewidth=2, label='Prediction')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$u(t,x)$')
    ax.set_title('$t = 0.25$', fontsize=10)
    ax.axis('square')
    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])

    # Slice at t = 0.50
    ax = plt.subplot(gs1[0, 1])
    ax.plot(x, Exact[50, :], 'b-', linewidth=2, label='Exact')
    ax.plot(x, U_pred[50, :], 'r--', linewidth=2, label='Prediction')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$u(t,x)$')
    ax.axis('square')
    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])
    ax.set_title('$t = 0.50$', fontsize=10)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.35), ncol=5, frameon=False)

    # Slice at t = 0.75
    ax = plt.subplot(gs1[0, 2])
    ax.plot(x, Exact[75, :], 'b-', linewidth=2, label='Exact')
    ax.plot(x, U_pred[75, :], 'r--', linewidth=2, label='Prediction')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$u(t,x)$')
    ax.axis('square')
    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])
    ax.set_title('$t = 0.75$', fontsize=10)

    # Save figure
    savefig("Burgers")
```

## PINN for a Boundary Layer Problem: Solution of Equation using Central Difference

```python
# ---------------------------------------------------------
# Solution of a boundary value problem using
# the Central Finite Difference Method
# ---------------------------------------------------------

# Import required libraries
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Set plotting style
plt.style.use('seaborn-poster')
%matplotlib inline

# Set font style for plots
matplotlib.rc('font', family='serif', serif='cm10')
matplotlib.rc('text', usetex=True)


# ---------------------------------------------------------
# Define problem parameters
# ---------------------------------------------------------

# Parameter in the differential equation
nu = 1.0 / 10**-3

# Number of subintervals (grid divisions)
n = 100

# Grid spacing:
# Domain is from x = -1 to x = 1, so total length = 2
h = (1 + 1) / n

# Generate grid points from -1 to 1
x = np.linspace(-1, 1, n + 1)


# ---------------------------------------------------------
# Construct the finite difference matrix A
# ---------------------------------------------------------

# Initialize the matrix A with zeros
A = np.zeros((n + 1, n + 1))

# Boundary conditions:
# At the first and last nodes, we directly impose the value of u
A[0, 0] = 1
A[n, n] = 1


# ---------------------------------------------------------
# Fill matrix entries for interior points
# ---------------------------------------------------------
# Using the central finite difference approximation:
#
# u''(x_i) ≈ (u_{i-1} - 2u_i + u_{i+1}) / h^2
#
# This creates a tridiagonal system
for i in range(1, n):
    A[i, i - 1] = 1
    A[i, i] = -(2 + nu * h**2)
    A[i, i + 1] = 1


# ---------------------------------------------------------
# Construct the right-hand side vector b
# ---------------------------------------------------------

# Initialize b
b = np.zeros(n + 1)

# Source term evaluated at each grid point
b = (np.exp(x)) * h * h * nu


# ---------------------------------------------------------
# Impose boundary conditions in vector b
# ---------------------------------------------------------
# These correspond to:
# u(-1) = 1
# u(1)  = 0
b[0] = 1
b[-1] = 0


# ---------------------------------------------------------
# Solve the linear system A y = b
# ---------------------------------------------------------
# y_act contains the numerical solution at all grid points
y_act = np.linalg.solve(A, b)


# ---------------------------------------------------------
# Plot the numerical solution
# ---------------------------------------------------------
gs = gridspec.GridSpec(1, 1)
ax = plt.subplot(gs[0, 0])

# Plot solution u versus x
x_fd, u_fd = x, y_act
ax.plot(x, y_act, "-", lw=2.0, color="b")


# Label axes
ax.set_xlabel('$x$')
ax.set_ylabel("$u$")

# Show the plot
plt.show()
```

```python
# =========================================================
# Solving a 1D Boundary Value Problem using a Vanilla PINN
# and comparing it with the Finite Difference (FD) solution
# =========================================================

# ---------------------------------------------------------
# Import required libraries
# ---------------------------------------------------------
import sys
sys.path.insert(0, 'Utilities/')

import os
import time
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import tensorflow as tf

from pyDOE import lhs
from mpl_toolkits.axes_grid1 import make_axes_locatable
from plotting import newfig, savefig


# ---------------------------------------------------------
# Reproducibility: fix random seeds
# ---------------------------------------------------------
np.random.seed(seed=1234)
tf.random.set_seed(1234)

# Disable TensorFloat-32 for more controlled numerics
tf.config.experimental.enable_tensor_float_32_execution(False)


# ---------------------------------------------------------
# Domain bounds
# ---------------------------------------------------------
lb = -1.0
ub = 1.0


# ---------------------------------------------------------
# Weight initialization function
# ---------------------------------------------------------
def hyper_initial(size):
    """
    Initialize the weights of one neural network layer.

    Parameters:
        size : list or tuple
            [input_dimension, output_dimension]

    Returns:
        tf.Variable
            Weight matrix initialized with truncated normal distribution
    """
    in_dim = size[0]
    out_dim = size[1]

    # Scale depends on number of input and output neurons
    std = np.sqrt(2.0 / (in_dim + out_dim))

    return tf.Variable(tf.random.truncated_normal(shape=size, stddev=std))


# ---------------------------------------------------------
# Neural network forward pass
# ---------------------------------------------------------
def DNN(X, W, b):
    """
    Forward propagation through the neural network.

    Parameters:
        X : input tensor
        W : list of weight matrices
        b : list of bias vectors

    Returns:
        Y : network output
    """

    # Normalize input from [lb, ub] to [-1, 1]
    A = 2.0 * (X - lb) / (ub - lb) - 1.0

    L = len(W)

    # Hidden layers with tanh activation
    for i in range(L - 1):
        A = tf.tanh(tf.add(tf.matmul(A, W[i]), b[i]))

    # Output layer with linear activation
    Y = tf.add(tf.matmul(A, W[-1]), b[-1])
    return Y


# ---------------------------------------------------------
# Helper function: collect all trainable variables
# ---------------------------------------------------------
def train_vars(W, b):
    """
    Return all trainable parameters (weights + biases).
    """
    return W + b


# ---------------------------------------------------------
# PINN solution u(x)
# ---------------------------------------------------------
def net_u(x, w, b):
    """
    Predict the solution u(x) using the neural network.
    """
    u = DNN(x, w, b)
    return u


# ---------------------------------------------------------
# PDE residual
# Governing equation:
#     u_xx - (1/nu)u - (1/nu)e^x = 0
# ---------------------------------------------------------
@tf.function
def net_f(x, W, b, nu):
    """
    Compute the physics residual using automatic differentiation.
    If the network satisfies the PDE perfectly, then f should be 0.
    """

    with tf.GradientTape(persistent=True) as tape1:
        tape1.watch([x])

        with tf.GradientTape(persistent=True) as tape2:
            tape2.watch([x])

            # Network prediction
            u = net_u(x, W, b)

        # First derivative du/dx
        u_x = tape2.gradient(u, x)
        del tape2

    # Second derivative d²u/dx²
    u_xx = tape1.gradient(u_x, x)
    del tape1

    # PDE residual
    f = u_xx - (1 / nu) * u - (1 / nu) * tf.exp(x)

    return f


# ---------------------------------------------------------
# One training step
# ---------------------------------------------------------
@tf.function
def train_step(W, b, X_u_train_tf, u_train_tf, X_f_train_tf, opt, nu):
    """
    Perform one optimization step.

    Total loss = boundary loss + physics loss
    """

    # Boundary points
    x_u = X_u_train_tf

    # Collocation points
    x_f = X_f_train_tf

    with tf.GradientTape() as tape:
        tape.watch([W, b])

        # Prediction at boundary points
        u_nn = net_u(x_u, W, b)

        # PDE residual at collocation points
        f_nn = net_f(x_f, W, b, nu)

        # Total loss:
        # Boundary condition fit + PDE residual fit
        loss = 100.0 * tf.reduce_mean(tf.square(u_nn - u_train_tf)) + \
               tf.reduce_mean(tf.square(f_nn))

    # Compute gradients
    grads = tape.gradient(loss, train_vars(W, b))

    # Update weights and biases
    opt.apply_gradients(zip(grads, train_vars(W, b)))

    return loss


# ---------------------------------------------------------
# Prediction function
# ---------------------------------------------------------
def predict(X_star_tf, w, b):
    """
    Predict u(x) at test points.
    """
    u_pred = net_u(X_star_tf, w, b)
    return u_pred


# ---------------------------------------------------------
# Finite Difference solution for the same problem
# ---------------------------------------------------------
def solve_fd(n, nu):
    """
    Solve the same boundary value problem using the
    central finite difference method.

    Governing equation:
        u_xx - (1/nu)u = (1/nu)e^x

    with boundary conditions:
        u(-1) = 1
        u(1)  = 0
    """

    # Grid spacing
    h = (ub - lb) / n

    # Grid points
    x_fd = np.linspace(lb, ub, n + 1)

    # Matrix A
    A = np.zeros((n + 1, n + 1))

    # Right-hand side vector
    rhs = np.zeros(n + 1)

    # Boundary conditions in matrix
    A[0, 0] = 1.0
    A[n, n] = 1.0

    # Interior finite difference equations
    for i in range(1, n):
        A[i, i - 1] = 1.0
        A[i, i] = -(2.0 + (h ** 2) / nu)
        A[i, i + 1] = 1.0

    # Interior right-hand side
    rhs = (h ** 2 / nu) * np.exp(x_fd)

    # Impose boundary values
    rhs[0] = 1.0
    rhs[-1] = 0.0

    # Solve linear system
    u_fd = np.linalg.solve(A, rhs)

    return x_fd, u_fd


# ---------------------------------------------------------
# Main program
# ---------------------------------------------------------
nu = 10**3          # same as 1000
noise = 0.0
N_f = 300           # number of collocation points
Nmax = 3000         # training iterations

# Network architecture
layers = [1, 4, 4, 4, 4, 4, 4, 1]
L = len(layers)

# Initialize weights and biases
W = [hyper_initial([layers[l - 1], layers[l]]) for l in range(1, L)]
b = [tf.Variable(tf.zeros([1, layers[l]])) for l in range(1, L)]


# ---------------------------------------------------------
# Boundary conditions
# ---------------------------------------------------------
x_0 = -1.0
x_1 = 1.0
u_0 = 1.0
u_1 = 0.0

# Training data from boundary points only
X_u_train = np.vstack([x_0, x_1])
u_train = np.vstack([u_0, u_1])

# Collocation points inside the domain using Latin Hypercube Sampling
X_f_train = lb + (ub - lb) * lhs(1, N_f)

# Test points where solution will be predicted
X_f_star = np.linspace(lb, ub, 200).reshape((-1, 1))


# ---------------------------------------------------------
# Convert NumPy arrays to TensorFlow tensors
# ---------------------------------------------------------
X_u_train_tf = tf.convert_to_tensor(X_u_train, dtype=tf.float32)
u_train_tf = tf.convert_to_tensor(u_train, dtype=tf.float32)
X_f_train_tf = tf.convert_to_tensor(X_f_train, dtype=tf.float32)
X_star_tf = tf.convert_to_tensor(X_f_star, dtype=tf.float32)


# ---------------------------------------------------------
# Optimizer
# ---------------------------------------------------------
lr = 5e-3
optimizer = tf.optimizers.Adam(learning_rate=lr)


# ---------------------------------------------------------
# Training loop
# ---------------------------------------------------------
start_time = time.time()
n = 0
loss = []

while n <= Nmax:
    loss_ = train_step(W, b, X_u_train_tf, u_train_tf, X_f_train_tf, optimizer, nu)
    loss.append(loss_)

    # Print progress every iteration
    if n%1000 == 0:
        print(f"Iteration is: {n} and loss is: {loss_}")

    n += 1

elapsed = time.time() - start_time
print('Training time: %.4f' % elapsed)


# ---------------------------------------------------------
# PINN prediction
# ---------------------------------------------------------
u_pred = predict(X_star_tf, W, b).numpy().flatten()
x_pred = X_f_star.flatten()


# ---------------------------------------------------------
# Finite Difference (FD) solution
# ---------------------------------------------------------
x_fd, u_fd = x_fd, u_fd


# ---------------------------------------------------------
# Compute relative error between PINN and FD solution
# ---------------------------------------------------------
# Interpolate FD solution onto PINN test points if needed
u_fd_interp = np.interp(x_pred, x_fd, u_fd)

error_rel = np.linalg.norm(u_fd_interp - u_pred, 2) / np.linalg.norm(u_fd_interp, 2)
print("Relative L2 error between PINN and FD solution: %e" % error_rel)


# ---------------------------------------------------------
# Plot 1: Compare PINN prediction and FD solution
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(x_fd, u_fd, 'b-', linewidth=2, label='FD solution (reference)')
plt.plot(x_pred, u_pred, 'r--', linewidth=2, label='PINN prediction')
plt.plot(X_u_train, u_train, 'ko', markersize=6, label='Boundary data')

plt.xlabel('$x$')
plt.ylabel('$u(x)$')
plt.title('Comparison of PINN and Finite Difference Solution')
plt.legend()
plt.grid(True)
plt.show()


# ---------------------------------------------------------
# Plot 2: Absolute error
# ---------------------------------------------------------
abs_error = np.abs(u_fd_interp - u_pred)

plt.figure(figsize=(8, 5))
plt.plot(x_pred, abs_error, 'm-', linewidth=2)
plt.xlabel('$x$')
plt.ylabel('Absolute Error')
plt.title('Absolute Error: |u_FD - u_PINN|')
plt.grid(True)
plt.show()


# ---------------------------------------------------------
# Plot 3: Training loss history
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(np.arange(len(loss)), [float(l.numpy()) for l in loss], linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('PINN Training Loss')
plt.yscale('log')
plt.grid(True)
plt.show()
```

# Self-Adaptive PINN for solving a 1D boundary value problem


```python
# =========================================================
# Self-Adaptive PINN for solving a 1D boundary value problem
# =========================================================

# ---------------------------------------------------------
# Import required libraries
# ---------------------------------------------------------
import sys
sys.path.insert(0, 'Utilities/')   # allow importing helper plotting functions

import os
import time
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from pyDOE import lhs                     # Latin Hypercube Sampling
from plotting import newfig, savefig
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable


# ---------------------------------------------------------
# Reproducibility
# ---------------------------------------------------------
np.random.seed(seed=1234)
tf.random.set_seed(1234)

# Disable TensorFloat32 for consistent numerical behavior
tf.config.experimental.enable_tensor_float_32_execution(False)


# ---------------------------------------------------------
# Domain boundaries
# ---------------------------------------------------------
lb = -1
ub = 1


# =========================================================
# Neural Network Initialization
# =========================================================
def hyper_initial(size):
    """
    Initialize weights for one neural network layer.

    Parameters
    ----------
    size : list
        [input_dimension, output_dimension]

    Returns
    -------
    tf.Variable
        weight matrix initialized using truncated normal distribution
    """

    in_dim = size[0]
    out_dim = size[1]

    # Xavier/He style scaling
    std = np.sqrt(2.0 / (in_dim + out_dim))

    return tf.Variable(
        tf.random.truncated_normal(shape=size, stddev=std)
    )


# =========================================================
# Neural Network Forward Pass
# =========================================================
def DNN(X, W, b):
    """
    Forward propagation through the neural network.
    """

    # Normalize input to [-1,1] (improves training stability)
    A = 2.0 * (X - lb) / (ub - lb) - 1.0

    L = len(W)

    # Hidden layers with tanh activation
    for i in range(L - 1):
        A = tf.tanh(tf.add(tf.matmul(A, W[i]), b[i]))

    # Output layer (linear)
    Y = tf.add(tf.matmul(A, W[-1]), b[-1])

    return Y


# =========================================================
# Helper functions for collecting trainable variables
# =========================================================

def train_vars_nn(W, b):
    """
    Return only neural network parameters.
    """
    return W + b


def train_vars_total(W, b, lambda_r, lambda_b):
    """
    Return all parameters including adaptive weights.
    """
    return W + b + lambda_r + lambda_b


def train_vars_sa(lambda_r, lambda_b):
    """
    Return only self-adaptive weights.
    """
    return lambda_r + lambda_b


# =========================================================
# Neural Network Solution
# =========================================================
def net_u(x, w, b):
    """
    Neural network prediction of solution u(x)
    """
    u = DNN(x, w, b)
    return u


# =========================================================
# Initialize self-adaptive weights
# =========================================================
def loss_weight(N_r, N_b):
    """
    Create adaptive weights for residual loss and boundary loss.

    N_r : number of residual points
    N_b : number of boundary points
    """

    # Boundary loss weights (start large)
    alpha_b = tf.Variable(
        tf.reshape(tf.repeat(1000.0, N_b), (N_b, -1))
    )

    # Residual loss weights
    alpha_r = tf.Variable(
        tf.ones(shape=[N_r, 1]), dtype=tf.float32
    )

    return alpha_r, alpha_b


# =========================================================
# PDE Residual using Automatic Differentiation
# =========================================================
@tf.function
def net_f(x, W, b, nu):
    """
    Compute PDE residual using automatic differentiation.

    PDE:
        u_xx - (1/nu)u - (1/nu)e^x = 0
    """

    with tf.GradientTape(persistent=True) as tape1:
        tape1.watch([x])

        with tf.GradientTape(persistent=True) as tape2:
            tape2.watch([x])

            u = net_u(x, W, b)

        # First derivative
        u_x = tape2.gradient(u, x)
        del tape2

    # Second derivative
    u_xx = tape1.gradient(u_x, x)
    del tape1

    # PDE residual
    f = u_xx - (1/nu) * u - (1/nu) * tf.exp(x)

    return f


# =========================================================
# Alternative PDE residual using finite difference
# =========================================================
@tf.function
def net_fd(x, W, b, nu):
    """
    Finite difference approximation of u_xx
    (for comparison with AD method)
    """

    h = 0.1

    u_xx = (1/(h**2)) * (
        net_u(x + h, W, b)
        - net_u(x, W, b)
        + net_u(x - h, W, b)
    )

    f = u_xx - (1/nu) * u - (1/nu) * tf.exp(x)

    return f


# =========================================================
# Training Step
# =========================================================
@tf.function()
def train_step(W, b, X_u_train_tf, u_train_tf, X_f_train_tf,
               opt, nu, lambda_r, lambda_b):

    # Boundary points
    x_u = X_u_train_tf

    # Residual points
    x_f = X_f_train_tf

    with tf.GradientTape(persistent=True) as tape:

        tape.watch([W, b, lambda_r, lambda_b])

        # NN prediction at boundary
        u_nn = net_u(x_u, W, b)

        # PDE residual
        f_nn = net_f(x_f, W, b, nu)

        # Weighted residual loss
        loss_r = tf.square(lambda_r * f_nn)

        # Weighted boundary loss
        loss_b = tf.square(lambda_b * (u_nn - u_train_tf))

        # Total loss
        loss = tf.reduce_mean(loss_b) + tf.reduce_mean(loss_r)

    # Gradients for network parameters
    grads = tape.gradient(loss, train_vars_nn(W, b))

    # Gradients for adaptive weights
    grads_u = tape.gradient(loss, lambda_r)
    grads_b = tape.gradient(loss, lambda_b)

    # Update neural network weights
    opt.apply_gradients(zip(grads, train_vars_nn(W, b)))

    # Update adaptive weights (gradient ascent)
    opt.apply_gradients(zip([-grads_u], [lambda_r]))
    opt.apply_gradients(zip([-grads_b], [lambda_b]))

    return loss


# =========================================================
# Prediction function
# =========================================================
def predict(X_star_tf, w, b):
    """
    Predict solution at test points
    """
    u_pred = net_u(X_star_tf, w, b)
    return u_pred


# =========================================================
# Problem setup
# =========================================================

nu = 10**-3        # PDE parameter
Nmax = 3000        # training iterations
N_f = 500          # residual points
N_b = 2            # boundary points


# Neural network architecture
layers = [1, 8, 8, 8, 8, 8, 8, 1]

L = len(layers)

# Initialize network parameters
W = [hyper_initial([layers[l-1], layers[l]]) for l in range(1, L)]
b = [tf.Variable(tf.zeros([1, layers[l]])) for l in range(1, L)]


# Initialize adaptive weights
alpha_r, alpha_b = loss_weight(N_f, N_b)


# =========================================================
# Boundary Conditions
# =========================================================

x_0 = -1
x_1 = 1

u_0 = 1
u_1 = 0

X_u_train = np.vstack([x_0, x_1])
u_train = np.vstack([u_0, u_1])


# =========================================================
# Collocation points inside domain
# =========================================================

X_f_train = lb + (ub - lb) * lhs(1, N_f)


# Test points for prediction
X_f_star = np.linspace(-1, 1, 200).reshape((-1, 1))


# Convert to TensorFlow tensors
X_u_train_tf = tf.convert_to_tensor(X_u_train, dtype=tf.float32)
u_train_tf = tf.convert_to_tensor(u_train, dtype=tf.float32)

X_f_train_tf = tf.convert_to_tensor(X_f_train, dtype=tf.float32)

X_star_tf = tf.convert_to_tensor(X_f_star, dtype=tf.float32)


# =========================================================
# Optimizer
# =========================================================

lr = 1e-3
optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=lr)




# =========================================================
# Training loop
# =========================================================

start_time = time.time()

n = 0
loss = []

while n <= Nmax:

    loss_ = train_step(
        W, b,
        X_u_train_tf,
        u_train_tf,
        X_f_train_tf,
        optimizer,
        nu,
        alpha_r,
        alpha_b
    )

    loss.append(loss_)

    if n%1000 == 0:
        print(f"Iteration is: {n} and loss is: {10**-7*loss_}")
    n += 1


elapsed = time.time() - start_time
print('Training time: %.4f' % (elapsed))


# ---------------------------------------------------------
# Plot 1: Compare PINN prediction and FD solution
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(x_fd, u_fd, 'b-', linewidth=2, label='FD solution (reference)')
plt.plot(x_pred, u_pred, 'r--', linewidth=2, label='PINN prediction')
plt.plot(X_u_train, u_train, 'ko', markersize=6, label='Boundary data')

plt.xlabel('$x$')
plt.ylabel('$u(x)$')
plt.title('Comparison of PINN and Finite Difference Solution')
plt.legend()
plt.grid(True)
plt.show()


# ---------------------------------------------------------
# Plot 2: Absolute error
# ---------------------------------------------------------
u_fd_interp = u_fd_interp.reshape(-1, 1)
abs_error = np.abs(u_fd_interp - u_pred)

plt.figure(figsize=(8, 5))
plt.plot(x_pred, abs_error, 'm-', linewidth=2)
plt.xlabel('$x$')
plt.ylabel('Absolute Error')
plt.title('Absolute Error: |u_FD - u_PINN|')
plt.grid(True)
plt.show()


# ---------------------------------------------------------
# Plot 3: Training loss history
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(np.arange(len(loss)), [float(l.numpy()) for l in loss], linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('PINN Training Loss')
plt.yscale('log')
plt.grid(True)
plt.show()

```

# Dynamic weight PINN for a 1D boundary value problem


```python
# =========================================================
# Dynamic weight PINN for a 1D boundary value problem
# =========================================================

# ---------------------------------------------------------
# Import required libraries
# ---------------------------------------------------------
import sys
sys.path.insert(0, 'Utilities/')   # allow access to helper plotting files

import os
import time
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from pyDOE import lhs   # Latin Hypercube Sampling


# ---------------------------------------------------------
# Reproducibility: fix random seeds
# ---------------------------------------------------------
np.random.seed(seed=1234)
tf.random.set_seed(1234)

# Disable TensorFloat-32 for more controlled numerics
tf.config.experimental.enable_tensor_float_32_execution(False)


# ---------------------------------------------------------
# Domain limits
# ---------------------------------------------------------
lb = -1
ub = 1


# =========================================================
# Weight initialization
# =========================================================
def hyper_initial(size):
    """
    Initialize weights for one neural network layer.

    Parameters:
        size : [input_dim, output_dim]

    Returns:
        tf.Variable containing initialized weights
    """
    in_dim = size[0]
    out_dim = size[1]

    # Xavier/He-like scaling
    std = np.sqrt(2.0 / (in_dim + out_dim))

    return tf.Variable(
        tf.random.truncated_normal(shape=size, stddev=std)
    )


# =========================================================
# Neural network forward pass
# =========================================================
def DNN(X, W, b):
    """
    Forward propagation through the neural network.

    Inputs are first normalized to [-1, 1].
    Hidden layers use tanh activation.
    Output layer is linear.
    """
    A = 2.0 * (X - lb) / (ub - lb) - 1.0
    L = len(W)

    # Hidden layers
    for i in range(L - 1):
        A = tf.tanh(tf.add(tf.matmul(A, W[i]), b[i]))

    # Output layer
    Y = tf.add(tf.matmul(A, W[-1]), b[-1])
    return Y


# ---------------------------------------------------------
# Collect trainable variables
# ---------------------------------------------------------
def train_vars(W, b):
    """
    Return all trainable neural network parameters.
    """
    return W + b


# =========================================================
# Neural network approximation u(x)
# =========================================================
def net_u(x, w, b):
    """
    Predict the solution u(x) using the neural network.
    """
    u = DNN(x, w, b)
    return u


# =========================================================
# PDE residual function
# Governing equation:
#     u_xx - (1/nu)u - (1/nu)e^x = 0
# =========================================================
@tf.function
def net_f(x, W, b, nu):
    """
    Compute the PDE residual using automatic differentiation.
    """

    with tf.GradientTape(persistent=True) as tape1:
        tape1.watch([x])

        with tf.GradientTape(persistent=True) as tape2:
            tape2.watch([x])

            # Network prediction
            u = net_u(x, W, b)

        # First derivative du/dx
        u_x = tape2.gradient(u, x)
        del tape2

    # Second derivative d²u/dx²
    u_xx = tape1.gradient(u_x, x)
    del tape1

    # PDE residual
    f = u_xx - (1 / nu) * u - (1 / nu) * tf.exp(x)
    return f


# =========================================================
# One training step
# =========================================================
@tf.function
def train_step(layers, W, b, X_u_train_tf, u_train_tf,
               X_f_train_tf, opt, nu, lambda_b, beta):
    """
    Perform one training step.

    This version adaptively updates the boundary-condition weight lambda_b
    based on the ratio of physics gradients to boundary gradients.
    """

    # Boundary points
    x_u = X_u_train_tf[:, 0:1]

    # Collocation points inside the domain
    x_f = X_f_train_tf[:, 0:1]

    # These lists will store temporary adaptive quantities
    adpative_constant_bcs_list = []
    lambda_b_list = []

    with tf.GradientTape(persistent=True) as tape:
        tape.watch([W, b])

        # Prediction at boundary points
        u_nn = net_u(x_u, W, b)

        # PDE residual at collocation points
        f_nn = net_f(x_f, W, b, nu)

        # Boundary loss
        bc_loss = tf.reduce_mean(tf.square(u_nn - u_train_tf))

        # Physics/residual loss
        phys_loss = tf.reduce_mean(tf.square(f_nn))

        # Weighted total loss
        loss = lambda_b[-1] * bc_loss + phys_loss

        # Normalize total loss
        loss = loss / (1 + lambda_b[-1])

        # Individual normalized parts
        p_loss = phys_loss / (1 + lambda_b[-1])
        b_loss = lambda_b[-1] * bc_loss / (1 + lambda_b[-1])

    # -----------------------------------------------------
    # Update neural network weights
    # -----------------------------------------------------
    grad_loss = tape.gradient(loss, train_vars(W, b))
    opt.apply_gradients(zip(grad_loss, train_vars(W, b)))

    # -----------------------------------------------------
    # Compute gradients of each loss part separately
    # -----------------------------------------------------
    grads_bc = tape.gradient(bc_loss, train_vars(W, b))
    grads_phys = tape.gradient(phys_loss, train_vars(W, b))

    # -----------------------------------------------------
    # Update adaptive boundary weight lambda_b
    # -----------------------------------------------------
    for i in range(len(layers) - 1):
        adpative_constant_bcs_list.append(
            tf.reduce_mean(tf.abs(grads_phys[i])) /
            tf.reduce_mean(tf.abs(grads_bc[i]))
        )

    # Average ratio across layers
    lambda_b_new = tf.reduce_mean(tf.stack(adpative_constant_bcs_list))

    # Exponential moving average update
    lambda_b_new = (1 - beta) * lambda_b[-1] + beta * lambda_b_new

    lambda_b_list.append(lambda_b_new)

    return loss, lambda_b_list, p_loss, b_loss


# =========================================================
# Prediction function
# =========================================================
def predict(X_star_tf, w, b):
    """
    Predict solution at test points.
    """
    u_pred = net_u(X_star_tf, w, b)
    return u_pred


# =========================================================
# Problem setup
# =========================================================
nu = 10**-3
noise = 0.0
N_f = 500
Nmax = 50000

# Neural network architecture
layers = [1, 8, 8, 8, 8, 8, 8, 1]
L = len(layers)

# Initialize weights and biases
W = [hyper_initial([layers[l - 1], layers[l]]) for l in range(1, L)]
b = [tf.Variable(tf.zeros([1, layers[l]])) for l in range(1, L)]


# ---------------------------------------------------------
# Boundary conditions
# ---------------------------------------------------------
x_0 = -1
x_1 = 1
u_0 = 1
u_1 = 0

# Boundary training points
X_u_train = np.vstack([x_0, x_1])
u_train = np.vstack([u_0, u_1])

# Residual/collocation points
X_f_train = lb + (ub - lb) * lhs(1, N_f)

# Test points for plotting
X_f_star = np.linspace(-1, 1, 200).reshape((-1, 1))


# ---------------------------------------------------------
# Convert to TensorFlow tensors
# ---------------------------------------------------------
X_u_train_tf = tf.convert_to_tensor(X_u_train, dtype=tf.float32)
u_train_tf = tf.convert_to_tensor(u_train, dtype=tf.float32)

X_f_train_tf = tf.convert_to_tensor(X_f_train, dtype=tf.float32)
X_star_tf = tf.convert_to_tensor(X_f_star, dtype=tf.float32)


# ---------------------------------------------------------
# Initialize adaptive boundary weight
# ---------------------------------------------------------
lam_b = np.array([1.0])
lam_b_tf = tf.convert_to_tensor(lam_b, dtype=tf.float32)

# Store lambda_b as a list so the newest value can be appended each step
lambda_b_list = [lam_b_tf]


# ---------------------------------------------------------
# Optimizer
# ---------------------------------------------------------
lr = 1e-3
optimizer = tf.optimizers.Adam(learning_rate=lr)


# =========================================================
# Training loop
# =========================================================
start_time = time.time()

n = 0
loss = []
bc_loss = []
phys_loss = []

# Smoothing factor for adaptive update
beta = 0.1

while n <= Nmax:
    loss_, lambda_b_out, phys_loss_, bc_loss_ = train_step(
        layers, W, b,
        X_u_train_tf, u_train_tf,
        X_f_train_tf, optimizer,
        nu, lambda_b_list, beta
    )

    # Update adaptive weight list
    lambda_b_list = lambda_b_out

    # Store losses
    loss.append(loss_)
    bc_loss.append(bc_loss_)
    phys_loss.append(phys_loss_)

    if n%4000==0:
        print(f"Iteration is: {n} and loss is: {loss_}")
    n += 1

elapsed = time.time() - start_time
print('Training time: %.4f' % elapsed)
print('Training time: %.4f' % elapsed)


# =========================================================
# Prediction after training
# =========================================================
def predict(X_star_tf, w, b):
    """
    Predict the solution at all test points.
    """
    x_star = X_star_tf
    u_pred = net_u(x_star, w, b)
    return u_pred


# Reference solution from finite difference code
u_star = y_act

# PINN prediction
u_pred = predict(X_star_tf, W, b)

# Relative L2 error
error_u = np.linalg.norm(u_star - u_pred, 2) / np.linalg.norm(u_star, 2)
print('Error u: %e' % error_u)


# =========================================================
# Plot 1: Exact solution vs PINN prediction
# =========================================================
gs1 = gridspec.GridSpec(1, 1)
ax = plt.subplot(gs1[0, 0])

# Exact/reference solution
ax.plot(x, y_act, 'b-', linewidth=2, label='Exact')

# PINN predicted solution
ax.plot(X_f_star, u_pred, '--r', linewidth=2, label='Prediction')

ax.set_xlabel('$x$')
ax.set_ylabel('$u(x)$')
plt.legend()

ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-3.1, 1.1])

plt.show()


# =========================================================
# Plot 2: Boundary loss and residual loss
# =========================================================

# Convert TensorFlow tensors to NumPy arrays for plotting
loss_list = [loss[i].numpy() for i in range(len(loss))]
bc_loss_list = [bc_loss[i].numpy() for i in range(len(bc_loss))]
phys_loss_list = [phys_loss[i].numpy() for i in range(len(phys_loss))]

gs2 = gridspec.GridSpec(1, 1)
ax = plt.subplot(gs2[0, 0])

# Iteration numbers
ep = np.arange(0, Nmax + 1, 1)

# Plot both losses on logarithmic scale
ax.semilogy(ep, bc_loss_list, 'g-', linewidth=2, label="Boundary Loss")
ax.semilogy(ep, phys_loss_list, 'r-', linewidth=2, label="Residual Loss")

ax.set_xlabel('Iteration')
ax.set_ylabel('Loss')

plt.legend()
plt.show()


# ---------------------------------------------------------
# Plot 1: Compare PINN prediction and FD solution
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(x_fd, u_fd, 'b-', linewidth=2, label='FD solution (reference)')
plt.plot(x_pred, u_pred, 'r--', linewidth=2, label='PINN prediction')
plt.plot(X_u_train, u_train, 'ko', markersize=6, label='Boundary data')

plt.xlabel('$x$')
plt.ylabel('$u(x)$')
plt.title('Comparison of PINN and Finite Difference Solution')
plt.legend()
plt.grid(True)
plt.show()


# ---------------------------------------------------------
# Plot 2: Absolute error
# ---------------------------------------------------------
u_fd_interp = u_fd_interp.reshape(-1, 1)
abs_error = np.abs(u_fd_interp - u_pred)

plt.figure(figsize=(8, 5))
plt.plot(x_pred, abs_error, 'm-', linewidth=2)
plt.xlabel('$x$')
plt.ylabel('Absolute Error')
plt.title('Absolute Error: |u_FD - u_PINN|')
plt.grid(True)
plt.show()


# ---------------------------------------------------------
# Plot 3: Training loss history
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(np.arange(len(loss)), [float(l.numpy()) for l in loss], linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('PINN Training Loss')
plt.yscale('log')
plt.grid(True)
plt.show()

```

# Hard Constraint PINN

```python
#https://github.com/jnh277/Linearly-Constrained-NN/blob/release/simulted_divergence_free.py

import torch
from matplotlib import pyplot as plt
from torch.utils import data
import numpy as np
import torch.autograd as ag
import torch.nn as nn
torch.manual_seed(1234)
epochs = 400
n_data = 200
### Input Data
def vector_field(x, y, a=0.01):
    v1 = torch.exp(-a*x*y)*(a*x*torch.sin(x*y) - x*torch.cos(x*y))
    v2 = torch.exp(-a*x*y)*(y*torch.cos(x*y) - a*y*torch.sin(x*y))
    return (v1, v2)


## ------------------ set up models-------------------------- ##
# set network size
n_in = 2
n_h1 = 100
n_h2 = 50
n_o = 1

# two outputs for the unconstrained network
n_o_uc = 2

# define model class
class DivFree2D(torch.nn.Module):
    def __init__(self, base_net):
        super(DivFree2D, self).__init__()
        self.base_net = base_net

    def forward(self, x):
        x.requires_grad = True
        y = self.base_net(x)
        dydx = ag.grad(outputs=y, inputs=x, create_graph=True, grad_outputs=torch.ones(y.size()),
                       retain_graph=True, only_inputs=True)[0]
        return y, dydx[:,1].unsqueeze(1), -dydx[:,0].unsqueeze(1)

model = DivFree2D(nn.Sequential(nn.Linear(n_in,n_h1),nn.Tanh(),nn.Linear(n_h1,n_h2),
                                         nn.Tanh(),nn.Linear(n_h2,n_o)))


model_uc = torch.nn.Sequential(
    torch.nn.Linear(n_in, n_h1),
    torch.nn.Tanh(),
    torch.nn.Linear(n_h1, n_h2),
    torch.nn.Tanh(),
    torch.nn.Linear(n_h2, n_o_uc),
)


# pregenerate validation data
x_val = 4.0 * torch.rand(2000, 2)
x1_val = x_val[:, 0].unsqueeze(1)
x2_val = x_val[:, 1].unsqueeze(1)

(v1, v2) = vector_field(x1_val, x2_val)
y1_val = v1 + 0.1 * torch.randn(x1_val.size())
y2_val = v2 + 0.1 * torch.randn(x1_val.size())
y_val = torch.cat((y1_val, y2_val), 1)


# Get the true function values on a grid
xv, yv = torch.meshgrid([torch.arange(0.0, 20.0) * 4.0 / 20.0, torch.arange(0.0, 20.0) * 4.0 / 20.0])
(v1, v2) = vector_field(xv, yv)

# generate training data
x_train = 4.0 * torch.rand(n_data, 2)
x1_train = x_train[:, 0].unsqueeze(1)
x2_train = x_train[:, 1].unsqueeze(1)

(v1_t, v2_t) = vector_field(x1_train, x2_train)
y1_train = v1_t + 0.1 * torch.randn(x1_train.size())
y2_train = v2_t + 0.1 * torch.randn(x1_train.size())


# define Dataset class
class Dataset(data.Dataset):
    'Characterizes a dataset for PyTorch'

    def __init__(self, x1, x2, y1, y2):
        'Initialization'
        self.x1 = x1
        self.x2 = x2
        self.y1 = y1
        self.y2 = y2

    def __len__(self):
        'Denotes the total number of samples'
        return len(self.x1)

    def __getitem__(self, index):
        'Generates one sample of data'
        # Select sample
        x1 = self.x1[index]
        x2 = self.x2[index]
        y1 = self.y1[index]
        y2 = self.y2[index]

        return x1, x2, y1, y2

training_set = Dataset(x1_train, x2_train, y1_train, y2_train)

# data loader Parameters
DL_params = {'batch_size': 100,
             'shuffle': True,
             'num_workers': 0,
             'pin_memory': False}
training_generator = data.DataLoader(training_set, **DL_params)


# ---------------  Set up and train the constrained model -------------------------------
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau\
                                    (optimizer, patience=10,
                                     min_lr=1e-10,
                                     factor=0.5, cooldown=15)

def train(epoch):
    model.train()
    total_loss = 0
    n_batches = 0
    for x1_train, x2_train, y1_train, y2_train in training_generator:
        optimizer.zero_grad()
        x_train = torch.cat((x1_train, x2_train), 1)
        (yhat, v1hat, v2hat) = model(x_train)
        loss = (criterion(y1_train, v1hat) + criterion(y2_train, v2hat)) / 2  # divide by 2 as it is a mean
        loss.backward()
        optimizer.step()
        total_loss += loss
        n_batches += 1
    return total_loss / n_batches

def eval(epoch):
    model.eval()
    # with torch.no_grad():
    (yhat, v1hat, v2hat) = model(x_val)
    loss = (criterion(y1_val, v1hat) + criterion(y2_val, v2hat)) / 2
    return loss.cpu()


train_loss = np.empty([epochs, 1])
val_loss = np.empty([epochs, 1])

print('Training Constrained NN')

for epoch in range(epochs):
    train_loss[epoch] = train(epoch).detach().numpy()
    v_loss = eval(epoch)
    scheduler.step(v_loss)
    val_loss[epoch] = v_loss.detach().numpy()
    print('Constrained NN: epoch: ', epoch, 'training loss ', train_loss[epoch], \
          'validation loss', val_loss[epoch])


# work out the rms error for this one
x_pred = torch.cat((xv.reshape(20 * 20, 1), yv.reshape(20 * 20, 1)), 1)
(f_pred, v1_pred, v2_pred) = model(x_pred)
error_new = torch.cat((v1.reshape(400, 1) - v1_pred.detach(), v2.reshape(400, 1) - v2_pred.detach()), 0)
rms_error = torch.sqrt(sum(error_new * error_new) / 800)

# ---------------  Set up and train the uncconstrained model -------------------------------
optimizer_uc = torch.optim.Adam(model_uc.parameters(), lr=0.01)
scheduler_uc = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_uc, patience=10,
                                                     min_lr=1e-10,
                                                    factor=0.5,
                                                    cooldown=15)

def train_uc(epoch):
    model_uc.train()
    total_loss = 0
    n_batches = 0
    for x1_train, x2_train, y1_train, y2_train in training_generator:
        optimizer_uc.zero_grad()
        x_train = torch.cat((x1_train, x2_train), 1)
        vhat = model_uc(x_train)
        y_train = torch.cat((y1_train, y2_train), 1)
        loss = criterion(y_train, vhat)
        loss.backward()
        optimizer_uc.step()
        total_loss += loss.cpu()
        n_batches += 1
    return total_loss / n_batches

def eval_uc(epoch):
    model_uc.eval()
    with torch.no_grad():
        (vhat) = model_uc(x_val)
        loss = criterion(y_val, vhat)
    return loss.cpu()


train_loss_uc = np.empty([epochs, 1])
val_loss_uc = np.empty([epochs, 1])


print('Training standard NN')

for epoch in range(epochs):
    train_loss_uc[epoch] = train_uc(epoch).detach().numpy()
    v_loss = eval_uc(epoch)
    scheduler_uc.step(v_loss)
    val_loss_uc[epoch] = v_loss.detach().numpy()
    print('Standard NN: epoch: ', epoch, 'training loss ', \
          train_loss_uc[epoch], 'validation loss', val_loss_uc[epoch])


# work out final rms error for unconstrainted net
(v_pred_uc) = model_uc(x_pred)
v1_pred_uc = v_pred_uc[:, 0]
v2_pred_uc = v_pred_uc[:, 1]

error_uc = torch.cat((v1.reshape(400) - v1_pred_uc.detach(), v2.reshape(400) - v2_pred_uc.detach()), 0)
rms_uc = torch.sqrt(sum(error_uc * error_uc) / 800)



print('Finished')
print("Final RMSE for constrained neural network: ",rms_error.item())
print("Final RMSE for standard neural network: ",rms_uc.item())

with torch.no_grad():
    # Initialize plot
    f, ax = plt.subplots(2, 2, figsize=(20, 16))
    # ax.pcolor(xv,yv,f_scalar)
    ax[0, 0].quiver(xv, yv, v1, v2)
    ax[0, 0].quiver(xv, yv, v1_pred.reshape(20, 20).detach(), v2_pred.reshape(20, 20).detach(), color='r')
    ax[0, 0].legend(['true', 'predicted'], fontsize=22)
    ax[0, 0].set_title('Constrained NN ', fontsize=22)
    ax[0, 0].tick_params(axis='x', labelsize=14 )
    ax[0, 0].tick_params(axis='y', labelsize=14 )


    ax[1, 0].plot(np.log(train_loss),lw=3.0)
    ax[1, 0].plot(np.log(val_loss), lw=3.0)
    # ax[1].plot(loss_save[1:epoch].log().detach().numpy())
    ax[1, 0].set_xlabel('training epoch',  fontsize=22)
    ax[1, 0].set_ylabel('Loss',  fontsize=22)
    ax[1, 0].legend(['training loss', 'val loss'], fontsize=22)
    ax[1, 0].tick_params(axis='x', labelsize=14 )
    ax[1, 0].tick_params(axis='y', labelsize=14 )

    ax[0, 1].quiver(xv, yv, v1, v2)
    ax[0, 1].quiver(xv, yv, v1_pred_uc.reshape(20, 20).detach(), v2_pred_uc.reshape(20, 20).detach(), color='r')
    ax[0, 1].legend(['true', 'predicted'], fontsize=22)
    ax[0, 1].set_title('Standard NN ',fontsize=22)
    ax[0, 1].tick_params(axis='x', labelsize=14 )
    ax[0,1].tick_params(axis='y', labelsize=14 )

    ax[1, 1].plot(np.log(train_loss_uc),lw=3.0)
    ax[1, 1].plot(np.log(val_loss_uc),lw=3.0)
    ax[1, 1].set_ylabel('Loss', fontsize=22)
    ax[1, 1].set_xlabel('training epoch', fontsize=22)
    ax[1, 1].legend(['training loss','val loss'], fontsize=22)
    ax[1, 1].tick_params(axis='x', labelsize=14)
    ax[1,1].tick_params(axis='y', labelsize=14 )
    plt.savefig("Figure_1.png", dpi=300)
    plt.show()
   
    

    # Initialize second plot
    f2, ax2 = plt.subplots(1, 3, figsize=(48, 10))
    Q = ax2[0].quiver(xv, yv, v1, v2, scale=None, scale_units='inches')
    Q._init()
    assert isinstance(Q.scale, float)
    ax2[0].quiver(x1_train, x2_train, y1_train, y2_train, scale=Q.scale, scale_units='inches', color='r')
    ax2[0].set_xlabel('$x_1$')
    ax2[0].set_ylabel('$x_2$')
    ax2[0].tick_params(axis='x', labelsize=36)
    ax2[0].tick_params(axis='y', labelsize=36)
    


    error_new = torch.cat((v1.reshape(400, 1) - v1_pred.detach(), v2.reshape(400, 1) - v2_pred.detach()), 0)
    rms_new = torch.sqrt(sum(error_new * error_new) / 800)

    ax2[1].quiver(xv, yv, v1 - v1_pred.reshape(20, 20).detach(), v2 - v2_pred.reshape(20, 20).detach(),
                  scale=Q.scale, scale_units='inches')
    ax2[1].set_xlabel('$x_1$')
    ax2[1].set_ylabel('$x_2$')
    ax2[1].set_title('Constrained Approach RMSE={0:.2f}'.format(rms_new.item()), fontsize=36)
    ax2[1].tick_params(axis='x', labelsize=36 )
    ax2[1].tick_params(axis='y', labelsize=36 )
    


    error_uc = torch.cat((v1.reshape(400) - v1_pred_uc.detach(), v2.reshape(400) - v2_pred_uc.detach()), 0)
    rms_uc = torch.sqrt(sum(error_uc * error_uc) / 800)

    ax2[2].quiver(xv, yv, v1 - v1_pred_uc.reshape(20, 20).detach(), v2 - v2_pred_uc.reshape(20, 20).detach(),
                  scale=Q.scale, scale_units='inches')
    ax2[2].set_xlabel('$x_1$')
    ax2[2].set_ylabel('$x_2$')
    ax2[2].set_title('Unconstrained NN RMSE={0:.2f}'.format(rms_uc.item()), fontsize=36)
    ax2[2].tick_params(axis='x', labelsize=36 )
    ax2[2].tick_params(axis='y', labelsize=36 )
    plt.savefig("Figure_2.png", dpi=300)
    plt.show()
```
